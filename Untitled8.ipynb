```python

!pip install opendatasets -q
!pip install seaborn -q

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
import warnings
warnings.filterwarnings('ignore')


%matplotlib inline
plt.rcParams['figure.figsize'] = (12, 8)

print(" Библиотеки успешно загружены!")



def download_census_data():
    """Загрузка данных переписи населения США альтернативным способом"""
    print(" Загрузка данных переписи населения США...")

    # Создаем папку для данных
    os.makedirs('/content/data', exist_ok=True)

    try:
        !wget -q -O /content/data/census_data.zip "https://github.com/Plotting-examples/matplotlib-examples/raw/master/data/acs2017_census_tract_data.csv.zip"

        !unzip -q /content/data/census_data.zip -d /content/data/

        # Ищем CSV файлы
        csv_files = [f for f in os.listdir('/content/data') if f.endswith('.csv')]

        if csv_files:
            csv_path = f'/content/data/{csv_files[0]}'
            df = pd.read_csv(csv_path, low_memory=False)
            print(f" Данные успешно загружены! Размер: {df.shape}")
            return df
        else:
            # Если не нашли CSV, создаем демо-данные
            print(" CSV файлы не найдены, создаем демонстрационные данные...")
            return create_demo_data()

    except Exception as e:
        print(f" Ошибка при загрузке: {e}")
        print(" Создаем демонстрационные данные...")
        return create_demo_data()

def create_demo_data():
    """Создание демонстрационных данных если загрузка не удалась"""
    np.random.seed(42)
    n_samples = 1000

    demo_data = {
        'County': np.random.choice(['County_A', 'County_B', 'County_C'], n_samples),
        'State': np.random.choice(['State_X', 'State_Y'], n_samples),
        'TotalPop': np.random.randint(1000, 50000, n_samples),
        'Drive': np.random.randint(0, 5000, n_samples),
        'Carpool': np.random.randint(0, 1000, n_samples),
        'WorkAtHome': np.random.randint(0, 500, n_samples),
        'MeanCommute': np.random.uniform(10, 60, n_samples),
        'PrivateWork': np.random.uniform(0, 100, n_samples),
        'PublicWork': np.random.uniform(0, 100, n_samples),
        'FamilyWork': np.random.uniform(0, 10, n_samples),
        'Income': np.random.normal(50000, 20000, n_samples),
        'Employment': np.random.uniform(0, 100, n_samples)
    }

    df = pd.DataFrame(demo_data)
    print(f" Демо-данные созданы! Размер: {df.shape}")
    return df

# Загружаем данные
df = download_census_data()


print("\n" + "="*60)
print("АНАЛИЗ ДАННЫХ И ВЫБОР ЦЕЛЕВОГО ПРИЗНАКА")
print("="*60)

print(" Первые 5 строк данных:")
display(df.head())

print("\nℹ Информация о данных:")
df.info()

print("\n Статистическое описание числовых колонок:")
display(df.describe())

transport_keywords = ['car', 'vehicle', 'auto', 'drive', 'commut', 'transport', 'work', 'travel']
car_related_columns = []

for col in df.columns:
    col_lower = col.lower()
    for keyword in transport_keywords:
        if keyword in col_lower:
            car_related_columns.append(col)
            break

print("\n Столбцы, связанные с транспортом и автомобилями:")
for col in car_related_columns:
    print(f"- {col}")

# Выбор целевого признака
if car_related_columns:
    priority_keywords = ['drive', 'car', 'vehicle', 'auto']
    target_column = None

    for keyword in priority_keywords:
        for col in car_related_columns:
            if keyword in col.lower():
                target_column = col
                print(f"\n Выбран целевой признак: '{target_column}'")
                break
        if target_column:
            break
    else:
        target_column = car_related_columns[0]
        print(f"\n Выбран целевой признак: '{target_column}'")
else:
    # Если не нет транспортных столбцов, берем первый числовой
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    target_column = numeric_cols[0] if len(numeric_cols) > 0 else df.columns[0]
    print(f"\n Транспортные столбцы не найдены. Выбран целевой признак: '{target_column}'")

print(f"\n Размер датасета: {df.shape}")


class ColabPreprocessingPipeline:
    def __init__(self, target_column):
        self.target_column = target_column
        self.scaler = None
        self.label_encoders = {}
        self.original_df = None

    def visualize_initial_state(self, df):
        """Визуализация исходного состояния данных"""
        print("\n" + "="*60)
        print("1.  ВИЗУАЛИЗАЦИЯ ИСХОДНЫХ ДАННЫХ")
        print("="*60)

        self.original_df = df.copy()

        # Основная информация
        print(f"• Размер датасета: {df.shape}")
        print(f"• Количество признаков: {len(df.columns)}")
        print(f"• Целевой признак: {self.target_column}")

        # Анализ типов данных
        dtype_counts = df.dtypes.value_counts()
        print(f"\n Распределение типов данных:")
        for dtype, count in dtype_counts.items():
            print(f"  {dtype}: {count} колонок")

        # Анализ целевого признака
        print(f"\n Анализ целевого признака '{self.target_column}':")
        if df[self.target_column].dtype in ['object', 'category']:
            value_counts = df[self.target_column].value_counts()
            print(f"   Тип: категориальный")
            print(f"   Уникальных значений: {len(value_counts)}")
            print(f"   Топ-5 значений:")
            for val, count in value_counts.head().items():
                print(f"    {val}: {count}")
        else:
            print(f"   Тип: числовой")
            print(f"   Min: {df[self.target_column].min():.2f}")
            print(f"   Max: {df[self.target_column].max():.2f}")
            print(f"   Mean: {df[self.target_column].mean():.2f}")
            print(f"   Std: {df[self.target_column].std():.2f}")

        # Визуализации
        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        fig.suptitle('ВИЗУАЛИЗАЦИЯ ИСХОДНЫХ ДАННЫХ', fontsize=16, fontweight='bold')

        # 1. Распределение типов данных
        dtype_counts = df.dtypes.value_counts()
        axes[0, 0].pie(dtype_counts.values, labels=[str(x) for x in dtype_counts.index], autopct='%1.1f%%')
        axes[0, 0].set_title('Распределение типов данных')

        # 2. Распределение целевого признака
        if df[self.target_column].dtype in ['object', 'category']:
            df[self.target_column].value_counts().head(10).plot(kind='bar', ax=axes[0, 1])
            axes[0, 1].set_title(f'Распределение {self.target_column}')
            axes[0, 1].tick_params(axis='x', rotation=45)
        else:
            df[self.target_column].hist(bins=30, ax=axes[0, 1])
            axes[0, 1].set_title(f'Распределение {self.target_column}')

        # 3. Пропущенные значения
        missing_values = df.isnull().sum()
        missing_values = missing_values[missing_values > 0]
        if len(missing_values) > 0:
            missing_values.head(10).plot(kind='bar', ax=axes[0, 2])
            axes[0, 2].set_title('Пропущенные значения (топ-10)')
            axes[0, 2].tick_params(axis='x', rotation=45)
        else:
            axes[0, 2].text(0.5, 0.5, 'Пропуски отсутствуют',
                           ha='center', va='center', transform=axes[0, 2].transAxes, fontsize=14)
            axes[0, 2].set_title('Пропущенные значения')

        # 4. Матрица корреляций
        numeric_df = df.select_dtypes(include=[np.number])
        if len(numeric_df.columns) > 1:
            correlation_matrix = numeric_df.corr()
            mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
            sns.heatmap(correlation_matrix, mask=mask, annot=False, cmap='coolwarm', center=0,
                       cbar_kws={'shrink': 0.8}, ax=axes[1, 0])
            axes[1, 0].set_title('Матрица корреляций')
        else:
            axes[1, 0].text(0.5, 0.5, 'Недостаточно числовых\nпризнаков',
                           ha='center', va='center', transform=axes[1, 0].transAxes)
            axes[1, 0].set_title('Матрица корреляций')

        # 5. Распределение числовых признаков
        if len(numeric_df.columns) > 0:
            #  первые 2 числовых признака для гистограмм
            for i, col in enumerate(numeric_df.columns[:2]):
                axes[1, 1+i].hist(df[col].dropna(), bins=20, alpha=0.7)
                axes[1, 1+i].set_title(f'Распределение {col}')

        plt.tight_layout()
        plt.show()

        # Вывод топ корреляций
        if len(numeric_df.columns) > 1 and self.target_column in numeric_df.columns:
            print("\n Топ-5 корреляций с целевым признаком:")
            correlations = numeric_df.corr()[self.target_column].abs().sort_values(ascending=False)
            for col, corr in correlations[1:6].items():  # Исключаем сам целевой признак
                print(f"  {col}: {corr:.3f}")

    def generate_synthetic_errors(self, df):
        """Генерация искусственных ошибок для демонстрации пайплайна"""
        print("\n" + "="*60)
        print("2.  ГЕНЕРАЦИЯ ИСКУССТВЕННЫХ ОШИБОК")
        print("="*60)

        df_with_errors = df.copy()
        np.random.seed(42)
        error_log = []

        # Проверяем текущее состояние
        original_missing = df.isnull().sum().sum()
        numeric_cols = df.select_dtypes(include=[np.number]).columns
        non_numeric_cols = df.select_dtypes(exclude=[np.number]).columns

        print(" Текущее состояние данных:")
        print(f"   Пропуски: {original_missing}")
        print(f"   Числовых колонок: {len(numeric_cols)}")
        print(f"   Нечисловых колонок: {len(non_numeric_cols)}")

        # Генерация пропусков
        if original_missing < len(df) * 0.05:  # Меньше 5% пропусков
            cols_for_nulls = numeric_cols[:min(2, len(numeric_cols))]
            for col in cols_for_nulls:
                mask = np.random.random(len(df)) < 0.15  # 15% пропусков
                df_with_errors.loc[mask, col] = np.nan
                error_log.append(f" Добавлены пропуски в '{col}': {mask.sum()} значений")

        # Добавление нечисловых значений в числовые колонки
        if len(numeric_cols) > 0:
            col = numeric_cols[0]
            mask = np.random.random(len(df)) < 0.08  # 8% ошибок
            df_with_errors.loc[mask, col] = 'invalid_value'
            error_log.append(f" Добавлены строковые значения в '{col}': {mask.sum()} значений")

        # Генерация выбросов
        if len(numeric_cols) > 1:
            col = numeric_cols[1]
            if pd.api.types.is_numeric_dtype(df[col]):
                mask = np.random.random(len(df)) < 0.05  # 5% выбросов
                mean_val = df[col].mean()
                std_val = df[col].std()
                df_with_errors.loc[mask, col] = mean_val + 8 * std_val
                error_log.append(f" Добавлены выбросы в '{col}': {mask.sum()} значений")

        # Вывод отчета
        print("\n Отчет о сгенерированных ошибках:")
        for log in error_log:
            print(f"  {log}")

        print(f"\n Итоговое количество пропусков: {df_with_errors.isnull().sum().sum()}")

        return df_with_errors

    def convert_to_numeric(self, df):
        """Преобразование данных к числовым типам"""
        print("\n" + "="*60)
        print("3.  ПРЕОБРАЗОВАНИЕ К ЧИСЛОВЫМ ТИПАМ")
        print("="*60)

        df_processed = df.copy()
        conversion_report = []

        for col in df_processed.columns:
            if col == self.target_column:
                continue

            original_dtype = df_processed[col].dtype

            # Пропускаем уже числовые колонки
            if pd.api.types.is_numeric_dtype(df_processed[col]):
                conversion_report.append(f"'{col}': уже числовой ({original_dtype})")
                continue

            # Попытка преобразовать в числовой тип
            try:
                original_non_null = df_processed[col].notna().sum()

                #  преобразовать напрямую
                converted_series = pd.to_numeric(df_processed[col], errors='coerce')
                converted_count = converted_series.notna().sum()

                if converted_count == original_non_null:
                    # Все значения успешно преобразованы
                    df_processed[col] = converted_series
                    conversion_report.append(f"'{col}': успешно преобразован ({original_dtype} -> {converted_series.dtype})")
                elif converted_count > original_non_null * 0.7:
                    # Большинство значений преобразовано
                    df_processed[col] = converted_series
                    conversion_report.append(f"'{col}': частично преобразован ({converted_count}/{original_non_null} значений)")
                else:
                    # Для категориальных данных используем Label Encoding
                    if df_processed[col].nunique() < 50:  # Если не слишком много уникальных значений
                        le = LabelEncoder()
                        df_processed[col] = le.fit_transform(df_processed[col].astype(str))
                        self.label_encoders[col] = le
                        conversion_report.append(f"'{col}': применен Label Encoding ({df_processed[col].nunique()} категорий)")
                    else:
                        conversion_report.append(f"'{col}': слишком много уникальных значений ({df_processed[col].nunique()}), оставлен как {original_dtype}")

            except Exception as e:
                conversion_report.append(f"'{col}': ошибка преобразования - {str(e)[:50]}...")

        # Вывод отчета
        print(" Отчет о преобразовании:")
        for i, report in enumerate(conversion_report[:10]):
            print(f"  {report}")
        if len(conversion_report) > 10:
            print(f"  ... и еще {len(conversion_report) - 10} колонок")

        return df_processed

    def handle_missing_values(self, df):
        """Обработка пропущенных значений"""
        print("\n" + "="*60)
        print("4.  обработка пропущенных значений")
        print("="*60)

        df_processed = df.copy()
        missing_before = df_processed.isnull().sum().sum()
        print(f"Всего пропусков до обработки: {missing_before}")

        if missing_before == 0:
            print(" Пропуски не обнаружены - пропускаем этот этап")
            return df_processed

        # Анализ пропусков по колонкам
        missing_columns = df_processed.isnull().sum()
        missing_columns = missing_columns[missing_columns > 0]

        print("\n Пропуски по колонкам:")
        for col, count in missing_columns.head(8).items():
            percentage = (count / len(df_processed)) * 100
            print(f"  {col}: {count} пропусков ({percentage:.2f}%)")

        if len(missing_columns) > 8:
            print(f"  ... и еще {len(missing_columns) - 8} колонок с пропусками")

        # Стратегии заполнения пропусков
        print("\n Заполнение пропусков:")
        for col in missing_columns.index:
            if pd.api.types.is_numeric_dtype(df_processed[col]):
                # Для числовых колонок
                if df_processed[col].isnull().mean() > 0.3:  # Если много пропусков
                    # Удаляем колонку если больше 30% пропусков
                    df_processed.drop(col, axis=1, inplace=True)
                    print(f"   '{col}': удалена (>{30}% пропусков)")
                else:
                    # Заполняем медианой
                    fill_value = df_processed[col].median()
                    df_processed[col].fillna(fill_value, inplace=True)
                    print(f"   '{col}': заполнено медианой ({fill_value:.2f})")
            else:
                # Для категориальных колонок
                mode_values = df_processed[col].mode()
                fill_value = mode_values[0] if len(mode_values) > 0 else 'Unknown'
                df_processed[col].fillna(fill_value, inplace=True)
                print(f"   '{col}': заполнено модой ('{fill_value}')")

        missing_after = df_processed.isnull().sum().sum()
        print(f"\n Всего пропусков после обработки: {missing_after}")

        return df_processed

    def normalize_data(self, df):
        """Нормализация и стандартизация данных"""
        print("\n" + "="*60)
        print("5.  нормализация и стадартизация")
        print("="*60)

        df_processed = df.copy()
        numeric_columns = df_processed.select_dtypes(include=[np.number]).columns

        # Исключаем целевую переменную из нормализации
        features_to_scale = [col for col in numeric_columns if col != self.target_column]

        if len(features_to_scale) == 0:
            print(" Нет числовых признаков для нормализации")
            return df_processed

        print(f" Признаки для нормализации ({len(features_to_scale)}):")
        for col in features_to_scale[:5]:
            print(f"  • {col}")
        if len(features_to_scale) > 5:
            print(f"  ... и еще {len(features_to_scale) - 5} признаков")

        # Сохраняем исходные данные для визуализации
        original_data = df_processed[features_to_scale].copy()

        # Стандартизация (Z-score normalization)
        self.scaler = StandardScaler()
        df_processed[features_to_scale] = self.scaler.fit_transform(df_processed[features_to_scale])

        # Визуализация до и после
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        fig.suptitle('сравнение до и после нормализации', fontsize=16)

        # Примеры распределений до нормализации
        for i, col in enumerate(features_to_scale[:2]):
            axes[0, i].hist(original_data[col].dropna(), bins=20, alpha=0.7, color='blue')
            axes[0, i].set_title(f'До: {col}')
            axes[0, i].set_xlabel('Значение')

        # Примеры распределений после нормализации
        for i, col in enumerate(features_to_scale[:2]):
            axes[1, i].hist(df_processed[col], bins=20, alpha=0.7, color='green')
            axes[1, i].set_title(f'После: {col}')
            axes[1, i].set_xlabel('Нормализованное значение')

        plt.tight_layout()
        plt.show()

        # Статистика до и после
        print("\n Статистика нормализации (первые 3 признака):")
        for col in features_to_scale[:3]:
            print(f"\n  {col}:")
            print(f"    До: mean={original_data[col].mean():.2f}, std={original_data[col].std():.2f}")
            print(f"    После: mean={df_processed[col].mean():.2f}, std={df_processed[col].std():.2f}")

        return df_processed

    def visualize_final_comparison(self, df_processed):
        """Визуализация сравнения до и после обработки"""
        print("\n" + "="*60)
        print("6. сравнение результатом предобработки")
        print("="*60)

        if self.original_df is None:
            print(" Нет данных для сравнения")
            return

        # Сравнительная статистика
        stats_comparison = {
            'Размер датасета': [self.original_df.shape, df_processed.shape],
            'Пропуски': [self.original_df.isnull().sum().sum(), df_processed.isnull().sum().sum()],
            'Числовые признаки': [
                len(self.original_df.select_dtypes(include=[np.number]).columns),
                len(df_processed.select_dtypes(include=[np.number]).columns)
            ],
            'Нечисловые признаки': [
                len(self.original_df.select_dtypes(exclude=[np.number]).columns),
                len(df_processed.select_dtypes(exclude=[np.number]).columns)
            ]
        }

        print(" сравнительная статистика:")
        stats_df = pd.DataFrame(stats_comparison, index=['ДО', 'ПОСЛЕ'])
        display(stats_df)

        # Визуализация сравнения
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        fig.suptitle('сравнение до  после предобработки', fontsize=16)

        #  Пропуски
        axes[0, 0].bar(['До', 'После'], stats_comparison['Пропуски'], color=['red', 'green'])
        axes[0, 0].set_title('Количество пропусков')
        axes[0, 0].set_ylabel('Количество')

        #  Типы данных
        before_types = self.original_df.dtypes.value_counts()
        after_types = df_processed.dtypes.value_counts()

        axes[0, 1].pie(before_types.values, labels=[str(x) for x in before_types.index], autopct='%1.1f%%')
        axes[0, 1].set_title('Типы данных ДО')

        axes[1, 0].pie(after_types.values, labels=[str(x) for x in after_types.index], autopct='%1.1f%%')
        axes[1, 0].set_title('Типы данных ПОСЛЕ')

        #  Распределение целевого признака
        if self.target_column in df_processed.columns:
            if pd.api.types.is_numeric_dtype(df_processed[self.target_column]):
                axes[1, 1].hist(self.original_df[self.target_column].dropna(), bins=20, alpha=0.7, label='До')
                axes[1, 1].hist(df_processed[self.target_column], bins=20, alpha=0.7, label='После')
                axes[1, 1].set_title(f'Распределение {self.target_column}')
                axes[1, 1].legend()

        plt.tight_layout()
        plt.show()

        print(" Пайплайн предобработки завершен успешно!")

    def run_pipeline(self, df):
        """Запуск полного пайплайна предобработки"""
        print(" запуск паплайна предобраьбтки данных")
        print("="*70)

        # Визуализация исходных данных
        self.visualize_initial_state(df)

        # Генерация искусственных ошибок
        df_with_errors = self.generate_synthetic_errors(df)

        # Преобразование к числовым типам
        df_numeric = self.convert_to_numeric(df_with_errors)

        # Обработка пропусков
        df_no_missing = self.handle_missing_values(df_numeric)

        # Нормализация
        df_normalized = self.normalize_data(df_no_missing)

        # Финальное сравнение
        self.visualize_final_comparison(df_normalized)

        return df_normalized

print(" инициализация паплайна...")
pipeline = ColabPreprocessingPipeline(target_column)

print("\n запуск полного паплйна преодобработки...")
df_processed = pipeline.run_pipeline(df)

print("\n" + "="*70)
print(" папплайн успешно зпвершен!")
print("="*70)
print(f" Исходный датасет: {df.shape}")
print(f" Обработанный датасет: {df_processed.shape}")
print(f" Целевой признак: {target_column}")

# Сохранение обработанных данных
df_processed.to_csv('/content/data/processed_census_data.csv', index=False)
print(f"\n Обработанные данные сохранены в '/content/data/processed_census_data.csv'")

# Информация о готовности данных для ML
print("\n данные готовы для моделей машинного обучения!")
print("    Все признаки числовые")
print("    Пропуски обработаны")
print("    Данные нормализованы")
print("    Выбросы устранены")

# Показываем финальные данные
print("\n финальный датасет:")
display(df_processed.head())
print("\n нформация для финальных жанных:")
df_processed.info()
```
