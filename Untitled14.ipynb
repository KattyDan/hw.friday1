```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Загружаем данные
import kagglehub
import os
path = kagglehub.dataset_download("patelris/crop-yield-prediction-dataset")
csv_files = [f for f in os.listdir(path) if f.lower().endswith('.csv')]
df = pd.read_csv(os.path.join(path, csv_files[0]))

# 1. Делаем ВСЁ числовым
df_num = df.copy()

# Каждый текстовый столбец -> числа
for col in df_num.columns:
    if df_num[col].dtype == 'object':
        le = LabelEncoder()
        df_num[col] = le.fit_transform(df_num[col].astype(str).fillna('Unknown'))

# Заполняем пропуски медианой
for col in df_num.columns:
    if df_num[col].isnull().any():
        df_num[col].fillna(df_num[col].median(), inplace=True)

# 2. Удаляем столбцы где все значения одинаковые
cols_to_drop = [col for col in df_num.columns if df_num[col].std() == 0]
df_num = df_num.drop(columns=cols_to_drop)

# 3. МАТРИЦА КОРРЕЛЯЦИИ (главное!)
corr_matrix = df_num.corr()

# 4. РИСУЕМ тепловую карту корреляции
plt.figure(figsize=(14, 12))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            square=True, linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.title('МАТРИЦА КОРРЕЛЯЦИИ', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# 5. ВЫВОДИМ ЧИСЛА
print("="*50)
print("МАТРИЦА КОРРЕЛЯЦИИ (первые 10x10):")
print("="*50)
print(corr_matrix.iloc[:10, :10].round(3))

# 6. ГИСТОГРАММА ТОЛЬКО ДЛЯ AREA CODE
print("\n" + "="*50)
print("ГИСТОГРАММА РАСПРЕДЕЛЕНИЯ AREA CODE:")
print("="*50)

# Проверяем, есть ли столбец 'Area Code' в данных
if 'Area Code' in df_num.columns:
    plt.figure(figsize=(10, 6))

    # Создаем гистограмму для Area Code
    plt.hist(df_num['Area Code'], bins=30, alpha=0.7, color='lightblue', edgecolor='black')

    # Настройки графика
    plt.title('Распределение Area Code', fontsize=16, fontweight='bold')
    plt.xlabel('Area Code (код региона)', fontsize=12)
    plt.ylabel('Количество записей', fontsize=12)
    plt.grid(axis='y', alpha=0.3)

    # Добавляем информацию о данных
    total_count = len(df_num['Area Code'])
    unique_codes = df_num['Area Code'].nunique()
    plt.text(0.02, 0.98, f'Всего записей: {total_count}\nУникальных кодов: {unique_codes}',
             transform=plt.gca().transAxes, fontsize=10,
             verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

    plt.tight_layout()
    plt.show()

    # Выводим статистику
    print(f"\nСтатистика по Area Code:")
    print(f"  Всего записей: {total_count}")
    print(f"  Уникальных кодов регионов: {unique_codes}")
    print(f"  Минимальный код: {df_num['Area Code'].min()}")
    print(f"  Максимальный код: {df_num['Area Code'].max()}")
    print(f"  Среднее значение: {df_num['Area Code'].mean():.2f}")
    print(f"  Медиана: {df_num['Area Code'].median()}")

else:
    print("Столбец 'Area Code' не найден в данных!")
    print("Доступные столбцы:", list(df_num.columns))
```

```python
# 2. ПРЕОБРАЗОВАНИЕ ДАННЫХ К ЧИСЛОВЫМ ТИПАМ
print("\n2. ПРЕОБРАЗОВАНИЕ ДАННЫХ К ЧИСЛОВЫМ ТИПАМ")

# Проверка нечисловых столбцов
non_numeric_cols = df.select_dtypes(exclude=['number']).columns.tolist()

if len(non_numeric_cols) == 0:
    print(" ВНИМАНИЕ: Все данные уже являются числовыми!")
    print("   Создаем искусственный категориальный признак для демонстрации...")

    np.random.seed(42)
    artificial_categories = np.random.choice(['Low', 'Medium', 'High'], size=len(df))
    df['Yield_Category'] = artificial_categories
    non_numeric_cols = ['Yield_Category']
    print(f"   Создан искусственный категориальный признак: {non_numeric_cols}")

print(f"Нечисловые столбцы для преобразования: {non_numeric_cols}")

# Преобразование категориальных признаков в числовые
for col in non_numeric_cols:
    if col in df.columns:
        unique_count = df[col].nunique()

        if unique_count <= 10:  # Малое количество уникальных значений - one-hot encoding
            dummies = pd.get_dummies(df[col], prefix=col)
            df = pd.concat([df, dummies], axis=1)
            df.drop(col, axis=1, inplace=True)
            print(f"   One-hot encoding для {col}: создано {len(dummies.columns)} новых признаков")
        else:  # Много уникальных значений - label encoding
            df[col] = pd.factorize(df[col])[0]
            print(f"   Label encoding для {col} (уникальных значений: {unique_count})")

print("Преобразование к числовым типам завершено!")
print(f"Новый размер датасета: {df.shape}")
```

```python
# 3. ОБРАБОТКА ПРОПУЩЕННЫХ ЗНАЧЕНИЙ
print("\n3. ОБРАБОТКА ПРОПУЩЕННЫХ ЗНАЧЕНИЙ")

# Проверка на наличие пропусков
missing_before = df.isnull().sum().sum()

if missing_before == 0:
    # ГЕНЕРАЦИЯ ОШИБКИ ЧЕРЕЗ LLM-СТИЛЬ
    print(" ВНИМАНИЕ: Пропущенные значения отсутствуют!")
    print("   Имитация проблемы: создаем искусственные пропуски...")

    # Создаем искусственные пропуски для демонстрации
    np.random.seed(42)

    # Выбираем столбцы для создания пропусков (включая целевой Value)
    cols_to_modify = []
    if 'Value' in df.columns:
        cols_to_modify.append('Value')

    # Добавляем другие числовые столбцы
    numeric_cols = df.select_dtypes(include='number').columns.tolist()
    other_cols = [col for col in numeric_cols if col not in cols_to_modify][:2]
    cols_to_modify.extend(other_cols)

    for col in cols_to_modify:
        mask = np.random.random(len(df)) < 0.1  # 10% пропусков
        df.loc[mask, col] = np.nan
        print(f"   Создано {mask.sum()} пропусков в столбце {col}")

# Проверяем пропуски после создания
missing_after = df.isnull().sum()
missing_cols = missing_after[missing_after > 0]

if len(missing_cols) > 0:
    print("\nСтолбцы с пропущенными значениями:")
    for col, count in missing_cols.items():
        print(f"   {col}: {count} пропусков ({count/len(df)*100:.1f}%)")

    # Стратегия заполнения пропусков
    for col in missing_cols.index:
        if df[col].dtype in ['float64', 'int64']:
            # Для числовых признаков
            if col == 'Value':  # Особый подход для целевого признака
                # Для целевого признака используем медиану
                fill_value = df[col].median()
                df[col].fillna(fill_value, inplace=True)
                print(f"   Заполнено пропусков в {col} (медиана): {missing_cols[col]}")
            else:
                # Для остальных числовых - среднее
                fill_value = df[col].mean()
                df[col].fillna(fill_value, inplace=True)
                print(f"   Заполнено пропусков в {col} (среднее): {missing_cols[col]}")
        else:
            # Для нечисловых - мода
            mode_value = df[col].mode()[0] if len(df[col].mode()) > 0 else 'Unknown'
            df[col].fillna(mode_value, inplace=True)
            print(f"   Заполнено пропусков в {col} (мода): {missing_cols[col]}")
else:
    print("Пропущенные значения отсутствуют")

print("\nОбработка пропусков завершена!")
```

```python
# 4. НОРМАЛИЗАЦИЯ И СТАНДАРТИЗАЦИЯ
print("\n4. НОРМАЛИЗАЦИЯ И СТАНДАРТИЗАЦИЯ")

# Определяем целевой признак для вашего датасета
target = 'Value'

# Проверка, нужна ли нормализация
numeric_columns_after = df.select_dtypes(include=['number']).columns
target_in_numeric = target in numeric_columns_after

if target_in_numeric:
    features_for_scaling = [col for col in numeric_columns_after if col != target]
else:
    features_for_scaling = numeric_columns_after.tolist()

print(f"Признаки для масштабирования: {len(features_for_scaling)}")

if len(features_for_scaling) == 0:
    # ГЕНЕРАЦИЯ ОШИБКИ ЧЕРЕЗ LLM-СТИЛЬ
    print(" ВНИМАНИЕ: Нет признаков для нормализации!")
    print("   Имитация проблемы: создаем искусственные признаки...")

    np.random.seed(42)

    # Создаем искусственные признаки с разными масштабами
    df['Feature_Large'] = np.random.uniform(1000, 10000, len(df))
    df['Feature_Small'] = np.random.uniform(0, 1, len(df))

    features_for_scaling = ['Feature_Large', 'Feature_Small']
    print(f"   Созданы признаки: {features_for_scaling}")

# Анализ распределения перед нормализацией
print("\nСтатистика до нормализации (первые 3 признака):")
for col in features_for_scaling[:3]:
    if col in df.columns:
        print(f"   {col}: mean={df[col].mean():.2f}, std={df[col].std():.2f}, min={df[col].min():.2f}, max={df[col].max():.2f}")

# Стандартизация (Z-score normalization)
scaler = StandardScaler()
df_scaled = pd.DataFrame(
    scaler.fit_transform(df[features_for_scaling]),
    columns=[f"{col}_standardized" for col in features_for_scaling],
    index=df.index
)

# Min-Max нормализация
minmax_scaler = MinMaxScaler()
df_minmax = pd.DataFrame(
    minmax_scaler.fit_transform(df[features_for_scaling]),
    columns=[f"{col}_normalized" for col in features_for_scaling],
    index=df.index
)

# Добавляем преобразованные признаки к исходному датасету
df = pd.concat([df, df_scaled, df_minmax], axis=1)

print("\nСтатистика после стандартизации (первые 3 признака):")
for col in features_for_scaling[:3]:
    standardized_col = f"{col}_standardized"
    if standardized_col in df.columns:
        print(f"   {standardized_col}: mean={df[standardized_col].mean():.2f}, std={df[standardized_col].std():.2f}")

print("\nНормализация завершена!")
print(f"Финальный размер датасета: {df.shape}")
```

```python
# 5. ВИЗУАЛИЗАЦИИ ПОСЛЕ ПРЕДОБРАБОТКИ
print("\n5. ВИЗУАЛИЗАЦИИ ПОСЛЕ ПРЕДОБРАБОТКИ")

# Определяем целевой признак
target = 'Value'

plt.figure(figsize=(15, 10))

# 1. Распределение целевой переменной после обработки
plt.subplot(2, 3, 1)
if target in df.columns:
    plt.hist(df[target], bins=20, alpha=0.7, color='purple', edgecolor='black')
    plt.title(f'Распределение {target} (после)')
    plt.xlabel(target)
    plt.ylabel('Частота')
else:
    plt.text(0.5, 0.5, f'Целевой признак {target}\nне найден',
             ha='center', va='center', transform=plt.gca().transAxes)
    plt.title(f'Распределение {target} (после)')

# 2. Корреляционная матрица после обработки
plt.subplot(2, 3, 2)
numeric_df_after = df.select_dtypes(include='number')
if not numeric_df_after.empty:
    correlation_matrix_after = numeric_df_after.corr()
    sns.heatmap(correlation_matrix_after, annot=False, cmap='coolwarm', center=0)
    plt.title('Матрица корреляций (после)')
else:
    plt.text(0.5, 0.5, 'Нет числовых признаков',
             ha='center', va='center', transform=plt.gca().transAxes)
    plt.title('Матрица корреляций (после)')

# 3. Визуализация стандартизированных признаков
# Находим стандартизированные признаки
standardized_cols = [col for col in df.columns if '_zscore' in col or '_standardized' in col or '_std' in col]
plt.subplot(2, 3, 3)

if len(standardized_cols) > 0:
    standardized_col = standardized_cols[0]
    plt.hist(df[standardized_col], bins=20, alpha=0.7, color='pink', edgecolor='black')
    plt.title(f'Стандартизированный признак')
    plt.xlabel(standardized_col)
    plt.ylabel('Частота')
else:
    plt.text(0.5, 0.5, 'Нет стандартизированных\nпризнаков',
             ha='center', va='center', transform=plt.gca().transAxes)
    plt.title('Стандартизированные признаки')

# 4. Проверка пропусков после обработки
plt.subplot(2, 3, 4)
missing_after = df.isnull().sum().sum()

if missing_after == 0:
    plt.text(0.5, 0.5, 'Пропуски отсутствуют',
             ha='center', va='center', transform=plt.gca().transAxes, fontsize=12)
    plt.title('Пропуски после обработки')
else:
    missing_data_after = df.isnull().sum()
    missing_data_after = missing_data_after[missing_data_after > 0]
    plt.bar(missing_data_after.index, missing_data_after.values, color='green', alpha=0.7)
    plt.title('Пропуски после обработки')
    plt.xticks(rotation=45)

# 5. Визуализация Min-Max нормализованных признаков
plt.subplot(2, 3, 5)
# Находим Min-Max нормализованные признаки
minmax_cols = [col for col in df.columns if '_minmax' in col or '_normalized' in col]

if len(minmax_cols) > 0:
    minmax_col = minmax_cols[0]
    plt.hist(df[minmax_col], bins=20, alpha=0.5, color='pink', edgecolor='black')
    plt.title('Min-Max нормализованный признак')
    plt.xlabel(minmax_col)
    plt.ylabel('Частота')
else:
    plt.text(0.5, 0.5, 'Нет Min-Max\nнормализованных признаков',
             ha='center', va='center', transform=plt.gca().transAxes)
    plt.title('Min-Max нормализация')

# 6. Boxplot целевой переменной
plt.subplot(2, 3, 6)
if target in df.columns:
    plt.boxplot(df[target].dropna())
    plt.title(f'Boxplot {target} (после)')
    plt.ylabel('Значение')
else:
    plt.text(0.5, 0.5, 'Нет данных\nдля boxplot',
             ha='center', va='center', transform=plt.gca().transAxes)
    plt.title('Boxplot анализ')

plt.tight_layout()
plt.show()
```
