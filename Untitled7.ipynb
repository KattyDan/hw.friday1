```python
# bibimyy_colab_fixed.py
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import SMOTE, ADASYN, RandomOverSampler
from imblearn.under_sampling import TomekLinks
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print("–£—Å—Ç–∞–Ω–æ–≤–∫–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫...")
!pip install imbalanced-learn -q

print("=" * 60)
print("–°–û–ó–î–ê–ù–ò–ï –î–ï–ú–û–ù–°–¢–†–ê–¶–ò–û–ù–ù–û–ì–û –î–ê–¢–ê–°–ï–¢–ê")
print("=" * 60)

# –°–æ–∑–¥–∞–µ–º —Ä–µ–∞–ª–∏—Å—Ç–∏—á–Ω—ã–π –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–æ–Ω–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç —Å —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω–æ–π —Ç–µ–º–∞—Ç–∏–∫–æ–π
np.random.seed(42)
n_samples = 2000

# –°–æ–∑–¥–∞–µ–º –¥–∞–Ω–Ω—ã–µ —Å —è–≤–Ω—ã–º –¥–∏—Å–±–∞–ª–∞–Ω—Å–æ–º –∫–ª–∞—Å—Å–æ–≤
data = {
    'age': np.random.randint(18, 75, n_samples),
    'income': np.random.normal(50000, 20000, n_samples),
    'drive_distance_km': np.random.normal(25, 12, n_samples),  # –û—Å–Ω–æ–≤–Ω–æ–π —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫
    'vehicle_count': np.random.randint(0, 4, n_samples),
    'work_commute_time': np.random.normal(45, 20, n_samples),
    'public_transport_usage': np.random.choice([0, 1, 2], n_samples, p=[0.6, 0.3, 0.1]),
    'car_ownership': np.random.choice([0, 1], n_samples, p=[0.3, 0.7]),
    'fuel_consumption': np.random.normal(8.5, 2.5, n_samples),
    'annual_mileage': np.random.normal(15000, 5000, n_samples),
    'traffic_congestion_hours': np.random.normal(10, 4, n_samples)
}

df = pd.DataFrame(data)

# –°–æ–∑–¥–∞–µ–º —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ —Å –¥–∏—Å–±–∞–ª–∞–Ω—Å–æ–º (3 –∫–ª–∞—Å—Å–∞)
# –ö–ª–∞—Å—Å 0: –ö–æ—Ä–æ—Ç–∫–∏–µ –ø–æ–µ–∑–¥–∫–∏ (60%)
# –ö–ª–∞—Å—Å 1: –°—Ä–µ–¥–Ω–∏–µ –ø–æ–µ–∑–¥–∫–∏ (30%)
# –ö–ª–∞—Å—Å 2: –î–ª–∏–Ω–Ω—ã–µ –ø–æ–µ–∑–¥–∫–∏ (10%)
df['commute_category'] = np.random.choice([0, 1, 2], n_samples, p=[0.6, 0.3, 0.1])

# –î–æ–±–∞–≤–ª—è–µ–º —Å–≤—è–∑—å –º–µ–∂–¥—É –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏ –∏ —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π
df.loc[df['commute_category'] == 0, 'drive_distance_km'] = np.random.normal(15, 5, np.sum(df['commute_category'] == 0))
df.loc[df['commute_category'] == 1, 'drive_distance_km'] = np.random.normal(25, 5, np.sum(df['commute_category'] == 1))
df.loc[df['commute_category'] == 2, 'drive_distance_km'] = np.random.normal(40, 8, np.sum(df['commute_category'] == 2))

print("–î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–æ–Ω–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç —Å–æ–∑–¥–∞–Ω —É—Å–ø–µ—à–Ω–æ!")
print(f"–†–∞–∑–º–µ—Ä –¥–∞—Ç–∞—Å–µ—Ç–∞: {df.shape}")
print(f"\n–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ 'commute_category':")
print(df['commute_category'].value_counts().sort_index())

print("\n–ü–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫ –¥–∞–Ω–Ω—ã—Ö:")
print(df.head())

# –ê–Ω–∞–ª–∏–∑ —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω—ã—Ö –∫–æ–ª–æ–Ω–æ–∫ (–∏–º–∏—Ç–∞—Ü–∏—è –≤–∞—à–µ–≥–æ –∫–æ–¥–∞)
print("\n" + "="*60)
print("–ê–ù–ê–õ–ò–ó –¢–†–ê–ù–°–ü–û–†–¢–ù–´–• –ö–û–õ–û–ù–û–ö")
print("="*60)

transport_keywords = ['car', 'vehicle', 'auto', 'drive', 'commute', 'transport', 'work', 'travel', 'distance', 'mileage', 'fuel', 'traffic']
car_related_columns = []

for col in df.columns:
    col_lower = col.lower()
    for keyword in transport_keywords:
        if keyword in col_lower:
            car_related_columns.append(col)
            break

print("\n–ö–æ–ª–æ–Ω–∫–∏, —Å–≤—è–∑–∞–Ω–Ω—ã–µ —Å —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–æ–º –∏ –∞–≤—Ç–æ–º–æ–±–∏–ª—è–º–∏:")
for col in car_related_columns:
    print(f'-- {col}')

# –í—ã–±–æ—Ä —Ü–µ–ª–µ–≤–æ–π –∫–æ–ª–æ–Ω–∫–∏
target_column = None
priority_keywords = ['drive', 'car', 'vehicle', 'auto', 'commute']
for keyword in priority_keywords:
    for col in car_related_columns:
        if keyword in col.lower():
            target_column = col
            print(f'\n–í—ã–±—Ä–∞–Ω —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫: {target_column}')
            break
    if target_column:
        break

if not target_column and car_related_columns:
    target_column = car_related_columns[0]
    print(f'\n–í—ã–±—Ä–∞–Ω —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ (–ø–µ—Ä–≤—ã–π –∏–∑ –¥–æ—Å—Ç—É–ø–Ω—ã—Ö): {target_column}')

# –ò—Å–ø–æ–ª—å–∑—É–µ–º —Å–æ–∑–¥–∞–Ω–Ω—ã–π —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫
target_column = 'commute_category'
print(f'\n–§–∏–Ω–∞–ª—å–Ω—ã–π —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫: {target_column}')

print("\n" + "="*60)
print("–ü–û–î–ì–û–¢–û–í–ö–ê –î–ê–ù–ù–´–• –ò –ú–û–î–ï–õ–ò–†–û–í–ê–ù–ò–ï")
print("="*60)

# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö
df_clean = df.copy()
y = df_clean[target_column]

print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤: {y.value_counts().to_dict()}")

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–ª–∞—Å—Å–æ–≤
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
y.value_counts().sort_index().plot(kind='bar', color=['red', 'blue', 'green'])
plt.title('–ò—Å—Ö–æ–¥–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤')
plt.xlabel('–ö–ª–∞—Å—Å')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
for i, count in enumerate(y.value_counts().sort_index()):
    plt.text(i, count + 10, str(count), ha='center', va='bottom')

# –í—ã–±–∏—Ä–∞–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∏ (–∏—Å–∫–ª—é—á–∞–µ–º —Ü–µ–ª–µ–≤—É—é)
numeric_columns = df_clean.select_dtypes(include=[np.number]).columns
feature_columns = [col for col in numeric_columns if col != target_column]

X = df_clean[feature_columns]

print(f"–ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è {len(feature_columns)} –ø—Ä–∏–∑–Ω–∞–∫–æ–≤: {feature_columns}")

# –†–∞–∑–¥–µ–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"\n–†–∞–∑–º–µ—Ä—ã –¥–∞–Ω–Ω—ã—Ö:")
print(f"X_train: {X_train.shape}, X_test: {X_test.shape}")
print(f"y_train —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: {pd.Series(y_train).value_counts().to_dict()}")

# 1) –ë–∞–∑–æ–≤–∞—è –º–æ–¥–µ–ª—å
print("\n" + "="*50)
print("1) –ë–ê–ó–û–í–ê–Ø –ú–û–î–ï–õ–¨ (–Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ)")
print("="*50)

base_model = DecisionTreeClassifier(random_state=42, max_depth=5)
base_model.fit(X_train, y_train)
y_pred_base = base_model.predict(X_test)
accuracy_base = accuracy_score(y_test, y_pred_base)

print(f"Accuracy: {accuracy_base:.4f}")
print("\n–û—Ç—á–µ—Ç –ø–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏:")
print(classification_report(y_test, y_pred_base))

# 2) –ú–æ–¥–µ–ª—å —Å –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–º –¥–∏—Å–±–∞–ª–∞–Ω—Å–æ–º
print("\n" + "="*50)
print("2) –ú–û–î–ï–õ–¨ –° –ò–°–ö–£–°–°–¢–í–ï–ù–ù–´–ú –î–ò–°–ë–ê–õ–ê–ù–°–û–ú")
print("="*50)

class_counts = pd.Series(y_train).value_counts()
majority_class = class_counts.idxmax()
majority_count = class_counts.max()

print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤ –≤ –æ–±—É—á–∞—é—â–∏—Ö –¥–∞–Ω–Ω—ã—Ö: {class_counts.to_dict()}")
print(f"–°–∞–º—ã–π –±–æ–ª—å—à–æ–π –∫–ª–∞—Å—Å: {majority_class} (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ: {majority_count})")

# –°–æ–∑–¥–∞–µ–º –¥–∏—Å–±–∞–ª–∞–Ω—Å - —É–º–µ–Ω—å—à–∞–µ–º –æ–¥–∏–Ω –∏–∑ –∫–ª–∞—Å—Å–æ–≤ –¥–æ 10% –æ—Ç majority
if len(class_counts) > 1:
    minority_class = class_counts.index[1]  # –í—Ç–æ—Ä–æ–π –ø–æ –≤–µ–ª–∏—á–∏–Ω–µ –∫–ª–∞—Å—Å
    minority_original_count = class_counts[minority_class]
    minority_indices = np.where(y_train == minority_class)[0]
    reduce_to_count = max(1, int(majority_count * 0.1))  # 10% –æ—Ç majority, –Ω–æ –Ω–µ –º–µ–Ω–µ–µ 1

    print(f"–£–º–µ–Ω—å—à–∞–µ–º –∫–ª–∞—Å—Å {minority_class} —Å {minority_original_count} –¥–æ {reduce_to_count} –ø—Ä–∏–º–µ—Ä–æ–≤")

    if len(minority_indices) > reduce_to_count:
        np.random.seed(42)
        reduced_indices = np.random.choice(minority_indices, size=reduce_to_count, replace=False)

        # –°–æ–∑–¥–∞–µ–º –Ω–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ
        keep_indices = np.where(y_train != minority_class)[0]
        imbalanced_indices = np.concatenate([keep_indices, reduced_indices])

        X_train_imbalanced = X_train.iloc[imbalanced_indices]
        y_train_imbalanced = y_train.iloc[imbalanced_indices]

        print(f"–ù–æ–≤–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: {pd.Series(y_train_imbalanced).value_counts().to_dict()}")

        # –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å
        imbalanced_model = DecisionTreeClassifier(random_state=42, max_depth=5)
        imbalanced_model.fit(X_train_imbalanced, y_train_imbalanced)
        y_pred_imbalanced = imbalanced_model.predict(X_test)
        accuracy_imbalanced = accuracy_score(y_test, y_pred_imbalanced)

        print(f"Accuracy: {accuracy_imbalanced:.4f}")
        print("\n–û—Ç—á–µ—Ç –ø–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏:")
        print(classification_report(y_test, y_pred_imbalanced))

        # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
        plt.subplot(1, 3, 2)
        pd.Series(y_train_imbalanced).value_counts().sort_index().plot(kind='bar', color=['red', 'blue', 'green'])
        plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ (–∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–π –¥–∏—Å–±–∞–ª–∞–Ω—Å)')
        plt.xlabel('–ö–ª–∞—Å—Å')
        plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
        for i, count in enumerate(pd.Series(y_train_imbalanced).value_counts().sort_index()):
            plt.text(i, count + 5, str(count), ha='center', va='bottom')

    else:
        print("–ù–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞")
        accuracy_imbalanced = None
else:
    print("–¢–æ–ª—å–∫–æ –æ–¥–∏–Ω –∫–ª–∞—Å—Å –≤ –¥–∞–Ω–Ω—ã—Ö")
    accuracy_imbalanced = None

# 3) –ú–µ—Ç–æ–¥—ã –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏
print("\n" + "="*50)
print("3) –ú–ï–¢–û–î–´ –ë–ê–õ–ê–ù–°–ò–†–û–í–ö–ò –ö–õ–ê–°–°–û–í")
print("="*50)

methods = {
    'RandomOverSampler': RandomOverSampler(random_state=42),
    'SMOTE': SMOTE(random_state=42),
    'ADASYN': ADASYN(random_state=42),
    'TomekLinks': TomekLinks()
}

results = {}

for method_name, sampler in methods.items():
    print(f"\n--- {method_name} ---")

    try:
        X_resampled, y_resampled = sampler.fit_resample(X_train, y_train)

        print(f"–†–∞–∑–º–µ—Ä –¥–æ –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏: {X_train.shape}")
        print(f"–†–∞–∑–º–µ—Ä –ø–æ—Å–ª–µ –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏: {X_resampled.shape}")
        print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–æ: {pd.Series(y_train).value_counts().to_dict()}")
        print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ—Å–ª–µ: {pd.Series(y_resampled).value_counts().to_dict()}")

        model = DecisionTreeClassifier(random_state=42, max_depth=5)
        model.fit(X_resampled, y_resampled)
        y_pred = model.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)

        results[method_name] = accuracy
        print(f"Accuracy: {accuracy:.4f}")

    except Exception as e:
        print(f"–û—à–∏–±–∫–∞: {e}")
        results[method_name] = None

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –º–µ—Ç–æ–¥–æ–≤
plt.subplot(1, 3, 3)
comparison_data = {'–ë–∞–∑–æ–≤–∞—è': accuracy_base}
if accuracy_imbalanced is not None:
    comparison_data['–î–∏—Å–±–∞–ª–∞–Ω—Å'] = accuracy_imbalanced
comparison_data.update(results)

colors = ['lightblue', 'red', 'green', 'orange', 'purple', 'brown']
bars = plt.bar(comparison_data.keys(), comparison_data.values(), color=colors[:len(comparison_data)])
plt.title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ accuracy –º–µ—Ç–æ–¥–æ–≤')
plt.xticks(rotation=45)
plt.ylabel('Accuracy')
plt.ylim(0, 1)

# –î–æ–±–∞–≤–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ —Å—Ç–æ–ª–±—Ü—ã
for bar, accuracy in zip(bars, comparison_data.values()):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
             f'{accuracy:.4f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# 4) –ò—Ç–æ–≥–æ–≤–æ–µ —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ
print("\n" + "="*50)
print("4) –°–†–ê–í–ù–ï–ù–ò–ï –†–ï–ó–£–õ–¨–¢–ê–¢–û–í")
print("="*50)

final_comparison = {'–ë–∞–∑–æ–≤–∞—è –º–æ–¥–µ–ª—å': accuracy_base}
if accuracy_imbalanced is not None:
    final_comparison['–ò—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–π –¥–∏—Å–±–∞–ª–∞–Ω—Å'] = accuracy_imbalanced
final_comparison.update(results)

comparison_df = pd.DataFrame({
    '–ú–µ—Ç–æ–¥': list(final_comparison.keys()),
    'Accuracy': list(final_comparison.values())
}).sort_values('Accuracy', ascending=False)

print("\n–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–µ—Ç–æ–¥–æ–≤ (–ø–æ —É–±—ã–≤–∞–Ω–∏—é accuracy):")
print(comparison_df.to_string(index=False))

print("\n" + "="*50)
print("–í–´–í–û–î–´ –ò –†–ï–ö–û–ú–ï–ù–î–ê–¶–ò–ò")
print("="*50)

best_method = comparison_df.iloc[0]['–ú–µ—Ç–æ–¥']
best_accuracy = comparison_df.iloc[0]['Accuracy']
worst_method = comparison_df.iloc[-1]['–ú–µ—Ç–æ–¥']
worst_accuracy = comparison_df.iloc[-1]['Accuracy']

print(f"‚úÖ –õ—É—á—à–∏–π –º–µ—Ç–æ–¥: {best_method} (Accuracy: {best_accuracy:.4f})")
print(f"‚ùå –•—É–¥—à–∏–π –º–µ—Ç–æ–¥: {worst_method} (Accuracy: {worst_accuracy:.4f})")

print("\nüìä –ê–Ω–∞–ª–∏–∑ –º–µ—Ç–æ–¥–æ–≤ –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏:")
print("1. RandomOverSampler - –ø—Ä–æ—Å—Ç–æ–µ –¥—É–±–ª–∏—Ä–æ–≤–∞–Ω–∏–µ –ø—Ä–∏–º–µ—Ä–æ–≤ –º–µ–Ω—å—à–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞")
print("2. SMOTE - —Å–æ–∑–¥–∞–µ—Ç —Å–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏–µ –ø—Ä–∏–º–µ—Ä—ã –Ω–∞ –æ—Å–Ω–æ–≤–µ –±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π –≤ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤")
print("3. ADASYN - –∞–¥–∞–ø—Ç–∏–≤–Ω—ã–π SMOTE, —É–¥–µ–ª—è–µ—Ç –±–æ–ª—å—à–µ –≤–Ω–∏–º–∞–Ω–∏—è —Å–ª–æ–∂–Ω—ã–º –¥–ª—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –ø—Ä–∏–º–µ—Ä–∞–º")
print("4. TomekLinks - undersampling –º–µ—Ç–æ–¥, —É–¥–∞–ª—è–µ—Ç –ø–æ–≥—Ä–∞–Ω–∏—á–Ω—ã–µ –ø—Ä–∏–º–µ—Ä—ã –¥–ª—è —É–ª—É—á—à–µ–Ω–∏—è —Ä–∞–∑–¥–µ–ª–∏–º–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–æ–≤")

print("\nüí° –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏:")
print("‚Ä¢ –ü—Ä–∏ —Å–∏–ª—å–Ω–æ–º –¥–∏—Å–±–∞–ª–∞–Ω—Å–µ –∫–ª–∞—Å—Å–æ–≤ —Ä–µ–∫–æ–º–µ–Ω–¥—É—é—Ç—Å—è SMOTE –∏–ª–∏ ADASYN")
print("‚Ä¢ RandomOverSampler –º–æ–∂–µ—Ç –ø—Ä–∏–≤–µ—Å—Ç–∏ –∫ –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏—é")
print("‚Ä¢ TomekLinks –ø–æ–ª–µ–∑–µ–Ω –∫–æ–≥–¥–∞ –º–Ω–æ–≥–æ –ø–µ—Ä–µ–∫—Ä—ã–≤–∞—é—â–∏—Ö—Å—è –ø—Ä–∏–º–µ—Ä–æ–≤ —Ä–∞–∑–Ω—ã—Ö –∫–ª–∞—Å—Å–æ–≤")
print("‚Ä¢ –í—Å–µ–≥–¥–∞ –æ—Ü–µ–Ω–∏–≤–∞–π—Ç–µ –Ω–µ —Ç–æ–ª—å–∫–æ accuracy, –Ω–æ –∏ precision/recall –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Å–∞")
print("‚Ä¢ –¢–µ—Å—Ç–∏—Ä—É–π—Ç–µ —Ä–∞–∑–Ω—ã–µ –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ –º–µ—Ç–æ–¥–æ–≤ –Ω–∞ –≤–∞—à–∏—Ö –¥–∞–Ω–Ω—ã—Ö")

print(f"\nüéØ –í –¥–∞–Ω–Ω–æ–º —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–µ –ª—É—á—à–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –ø–æ–∫–∞–∑–∞–ª –º–µ—Ç–æ–¥: {best_method}")
print("–≠—Ç–æ –¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É–µ—Ç —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –º–µ—Ç–æ–¥–æ–≤ –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏ –¥–ª—è —É–ª—É—á—à–µ–Ω–∏—è –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–µ–π –Ω–∞ –Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö!")
```

```python
# bibimyy_colab_fixed_final.py
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import SMOTE, ADASYN, RandomOverSampler
from imblearn.under_sampling import TomekLinks
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print("–£—Å—Ç–∞–Ω–æ–≤–∫–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫...")
!pip install imbalanced-learn -q

print("=" * 60)
print("–°–û–ó–î–ê–ù–ò–ï –î–ê–¢–ê–°–ï–¢–ê US CENSUS DEMOGRAPHIC DATA")
print("=" * 60)

# –°–æ–∑–¥–∞–µ–º –ø–∞–ø–∫—É –¥–ª—è –¥–∞–Ω–Ω—ã—Ö
if not os.path.exists('./data'):
    os.makedirs('./data')

# –°–æ–∑–¥–∞–µ–º —Ä–µ–∞–ª–∏—Å—Ç–∏—á–Ω—ã–π –¥–µ–º–æ-–¥–∞—Ç–∞—Å–µ—Ç –Ω–∞ –æ—Å–Ω–æ–≤–µ US Census —Å —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω–æ–π —Ç–µ–º–∞—Ç–∏–∫–æ–π
np.random.seed(42)
n_samples = 1500

print("–°–æ–∑–¥–∞–Ω–∏–µ —Ä–µ–∞–ª–∏—Å—Ç–∏—á–Ω–æ–≥–æ –¥–µ–º–æ-–¥–∞—Ç–∞—Å–µ—Ç–∞ US Census...")

# –°–æ–∑–¥–∞–µ–º –¥–∞–Ω–Ω—ã–µ —Å —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω–æ–π —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–æ–π
demo_data = {
    'State': np.random.choice(['California', 'Texas', 'New York', 'Florida', 'Illinois', 'Ohio'], n_samples),
    'County': np.random.choice(['Los Angeles', 'Cook', 'Harris', 'Maricopa', 'San Diego', 'Miami-Dade'], n_samples),
    'TotalPop': np.random.randint(50000, 5000000, n_samples),

    # –¢—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω—ã–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏
    'DriveAlone': np.random.randint(10000, 400000, n_samples),
    'Carpool': np.random.randint(5000, 150000, n_samples),
    'PublicTransport': np.random.randint(1000, 100000, n_samples),
    'Walk': np.random.randint(500, 50000, n_samples),
    'Bicycle': np.random.randint(100, 20000, n_samples),
    'WorkAtHome': np.random.randint(1000, 80000, n_samples),

    # –í—Ä–µ–º—è –≤ –ø—É—Ç–∏
    'MeanCommuteMinutes': np.random.uniform(15, 60, n_samples),

    # –î–µ–º–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏
    'IncomePerCapita': np.random.randint(25000, 80000, n_samples),
    'Employed': np.random.randint(20000, 300000, n_samples),
    'Unemployed': np.random.randint(1000, 30000, n_samples),

    # –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
    'Poverty': np.random.uniform(5, 25, n_samples),
    'ChildPoverty': np.random.uniform(8, 35, n_samples),
    'Professional': np.random.uniform(20, 60, n_samples),
    'Service': np.random.uniform(10, 40, n_samples),
    'Office': np.random.uniform(15, 45, n_samples),
    'Construction': np.random.uniform(5, 25, n_samples),
    'Production': np.random.uniform(8, 30, n_samples)
}

df = pd.DataFrame(demo_data)

# –°–æ–∑–¥–∞–µ–º —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ –Ω–∞ –æ—Å–Ω–æ–≤–µ –≤—Ä–µ–º–µ–Ω–∏ –≤ –ø—É—Ç–∏ (–∑–∞–¥–∞—á–∞ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏)
df['CommuteCategory'] = pd.cut(df['MeanCommuteMinutes'],
                              bins=[0, 25, 40, 100],
                              labels=[0, 1, 2])  # –ò—Å–ø–æ–ª—å–∑—É–µ–º —á–∏—Å–ª–æ–≤—ã–µ –º–µ—Ç–∫–∏

print("–î–µ–º–æ-–¥–∞—Ç–∞—Å–µ—Ç —É—Å–ø–µ—à–Ω–æ —Å–æ–∑–¥–∞–Ω!")
print(f"–†–∞–∑–º–µ—Ä –¥–∞—Ç–∞—Å–µ—Ç–∞: {df.shape}")

print("\n–ü–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫ –¥–∞–Ω–Ω—ã—Ö:")
print(df.head())

print("\n" + "="*60)
print("–ê–ù–ê–õ–ò–ó –¢–†–ê–ù–°–ü–û–†–¢–ù–´–• –ö–û–õ–û–ù–û–ö")
print("="*60)

# –ò—â–µ–º –∫–æ–ª–æ–Ω–∫–∏, —Å–≤—è–∑–∞–Ω–Ω—ã–µ —Å —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–æ–º –∏ –ø–µ—Ä–µ–º–µ—â–µ–Ω–∏—è–º–∏
transport_keywords = [
    'car', 'vehicle', 'auto', 'drive', 'commute', 'transport',
    'travel', 'distance', 'time', 'minutes', 'hours', 'work',
    'job', 'employment', 'alone', 'carpool', 'bus', 'train',
    'walk', 'bicycle', 'motorcycle', 'transit', 'public'
]

car_related_columns = []

for col in df.columns:
    col_lower = str(col).lower()
    for keyword in transport_keywords:
        if keyword in col_lower:
            car_related_columns.append(col)
            break

print("\n–ö–æ–ª–æ–Ω–∫–∏, —Å–≤—è–∑–∞–Ω–Ω—ã–µ —Å —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–æ–º –∏ –ø–µ—Ä–µ–º–µ—â–µ–Ω–∏—è–º–∏:")
for col in car_related_columns:
    print(f'-- {col}')

# –í—ã–±–æ—Ä —Ü–µ–ª–µ–≤–æ–π –∫–æ–ª–æ–Ω–∫–∏
target_column = 'CommuteCategory'
print(f'\n–¶–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫: {target_column}')

print("\n" + "="*60)
print("–ü–û–î–ì–û–¢–û–í–ö–ê –î–ê–ù–ù–´–• –ò –ú–û–î–ï–õ–ò–†–û–í–ê–ù–ò–ï")
print("="*60)

# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö
df_clean = df.copy()

# –¶–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è
y = df_clean[target_column].astype(int)  # –Ø–≤–Ω–æ –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ int

print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤: {y.value_counts().to_dict()}")

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–ª–∞—Å—Å–æ–≤
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
class_counts = y.value_counts().sort_index()
bars = plt.bar(class_counts.index, class_counts.values, color=['red', 'blue', 'green'])
plt.title('–ò—Å—Ö–æ–¥–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤')
plt.xlabel('–ö–ª–∞—Å—Å')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
plt.xticks([0, 1, 2])

# –î–æ–±–∞–≤–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ —Å—Ç–æ–ª–±—Ü—ã
for bar, count in zip(bars, class_counts.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 10,
             str(int(count)), ha='center', va='bottom')

# –í—ã–±–∏—Ä–∞–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∏ (–∏—Å–∫–ª—é—á–∞–µ–º —Ü–µ–ª–µ–≤—É—é –∏ –Ω–µ—á–∏—Å–ª–æ–≤—ã–µ –∫–æ–ª–æ–Ω–∫–∏)
numeric_columns = df_clean.select_dtypes(include=[np.number]).columns
exclude_cols = [target_column, 'State', 'County']  # –ò—Å–∫–ª—é—á–∞–µ–º —Ü–µ–ª–µ–≤—É—é –∏ —Ç–µ–∫—Å—Ç–æ–≤—ã–µ –∫–æ–ª–æ–Ω–∫–∏

feature_columns = [col for col in numeric_columns if col not in exclude_cols]

print(f"\n–ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è {len(feature_columns)} —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:")
for col in feature_columns:
    print(f"  - {col}")

X = df_clean[feature_columns].fillna(df_clean[feature_columns].mean())

# –†–∞–∑–¥–µ–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"\n–†–∞–∑–º–µ—Ä—ã –¥–∞–Ω–Ω—ã—Ö:")
print(f"X_train: {X_train.shape}, X_test: {X_test.shape}")
print(f"y_train —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: {y_train.value_counts().to_dict()}")

# 1) –ë–∞–∑–æ–≤–∞—è –º–æ–¥–µ–ª—å
print("\n" + "="*50)
print("1) –ë–ê–ó–û–í–ê–Ø –ú–û–î–ï–õ–¨ (–Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ)")
print("="*50)

base_model = DecisionTreeClassifier(random_state=42, max_depth=5)
base_model.fit(X_train, y_train)
y_pred_base = base_model.predict(X_test)
accuracy_base = accuracy_score(y_test, y_pred_base)

print(f"Accuracy: {accuracy_base:.4f}")
print("\n–û—Ç—á–µ—Ç –ø–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏:")
print(classification_report(y_test, y_pred_base))

# 2) –ú–æ–¥–µ–ª—å —Å –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–º –¥–∏—Å–±–∞–ª–∞–Ω—Å–æ–º
print("\n" + "="*50)
print("2) –ú–û–î–ï–õ–¨ –° –ò–°–ö–£–°–°–¢–í–ï–ù–ù–´–ú –î–ò–°–ë–ê–õ–ê–ù–°–û–ú")
print("="*50)

class_counts = y_train.value_counts()
majority_class = class_counts.idxmax()
majority_count = class_counts.max()

print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤ –≤ –æ–±—É—á–∞—é—â–∏—Ö –¥–∞–Ω–Ω—ã—Ö: {class_counts.to_dict()}")
print(f"–°–∞–º—ã–π –±–æ–ª—å—à–æ–π –∫–ª–∞—Å—Å: {majority_class} (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ: {majority_count})")

accuracy_imbalanced = None
if len(class_counts) > 1:
    # –í—ã–±–∏—Ä–∞–µ–º –º–µ–Ω—å—à–∏–Ω—Å—Ç–≤–µ–Ω–Ω—ã–π –∫–ª–∞—Å—Å –¥–ª—è —É–º–µ–Ω—å—à–µ–Ω–∏—è (–∫–ª–∞—Å—Å 1)
    minority_class = 1
    minority_original_count = class_counts.get(minority_class, 0)

    if minority_original_count > 0:
        minority_indices = y_train[y_train == minority_class].index
        reduce_to_count = max(1, int(majority_count * 0.1))  # 10% –æ—Ç majority

        print(f"–£–º–µ–Ω—å—à–∞–µ–º –∫–ª–∞—Å—Å {minority_class} —Å {minority_original_count} –¥–æ {reduce_to_count} –ø—Ä–∏–º–µ—Ä–æ–≤")

        if len(minority_indices) > reduce_to_count:
            np.random.seed(42)
            reduced_indices = np.random.choice(minority_indices, size=reduce_to_count, replace=False)

            # –°–æ–∑–¥–∞–µ–º –Ω–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ
            keep_indices = y_train[y_train != minority_class].index
            imbalanced_indices = keep_indices.union(reduced_indices)

            X_train_imbalanced = X_train.loc[imbalanced_indices]
            y_train_imbalanced = y_train.loc[imbalanced_indices]

            new_distribution = y_train_imbalanced.value_counts().to_dict()
            print(f"–ù–æ–≤–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: {new_distribution}")

            # –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å
            imbalanced_model = DecisionTreeClassifier(random_state=42, max_depth=5)
            imbalanced_model.fit(X_train_imbalanced, y_train_imbalanced)
            y_pred_imbalanced = imbalanced_model.predict(X_test)
            accuracy_imbalanced = accuracy_score(y_test, y_pred_imbalanced)

            print(f"Accuracy: {accuracy_imbalanced:.4f}")
            print("\n–û—Ç—á–µ—Ç –ø–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏:")
            print(classification_report(y_test, y_pred_imbalanced))

            # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
            plt.subplot(1, 3, 2)
            imbalanced_counts = y_train_imbalanced.value_counts().sort_index()
            bars2 = plt.bar(imbalanced_counts.index, imbalanced_counts.values, color=['red', 'blue', 'green'])
            plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ (–∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–π –¥–∏—Å–±–∞–ª–∞–Ω—Å)')
            plt.xlabel('–ö–ª–∞—Å—Å')
            plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
            plt.xticks([0, 1, 2])

            for bar, count in zip(bars2, imbalanced_counts.values):
                plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5,
                         str(int(count)), ha='center', va='bottom')

        else:
            print("–ù–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞")
    else:
        print("–í—ã–±—Ä–∞–Ω–Ω—ã–π –º–∏–Ω–æ—Ä–∏—Ç–∞—Ä–Ω—ã–π –∫–ª–∞—Å—Å –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç –≤ –¥–∞–Ω–Ω—ã—Ö")
else:
    print("–¢–æ–ª—å–∫–æ –æ–¥–∏–Ω –∫–ª–∞—Å—Å –≤ –¥–∞–Ω–Ω—ã—Ö")

# 3) –ú–µ—Ç–æ–¥—ã –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏
print("\n" + "="*50)
print("3) –ú–ï–¢–û–î–´ –ë–ê–õ–ê–ù–°–ò–†–û–í–ö–ò –ö–õ–ê–°–°–û–í")
print("="*50)

methods = {
    'RandomOverSampler': RandomOverSampler(random_state=42),
    'SMOTE': SMOTE(random_state=42),
    'ADASYN': ADASYN(random_state=42, sampling_strategy='auto'),
    'TomekLinks': TomekLinks()
}

results = {}

for method_name, sampler in methods.items():
    print(f"\n--- {method_name} ---")

    try:
        X_resampled, y_resampled = sampler.fit_resample(X_train, y_train)

        print(f"–†–∞–∑–º–µ—Ä –¥–æ: {X_train.shape}, –ø–æ—Å–ª–µ: {X_resampled.shape}")
        print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ—Å–ª–µ: {pd.Series(y_resampled).value_counts().to_dict()}")

        model = DecisionTreeClassifier(random_state=42, max_depth=5)
        model.fit(X_resampled, y_resampled)
        y_pred = model.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)

        results[method_name] = accuracy
        print(f"Accuracy: {accuracy:.4f}")

    except Exception as e:
        print(f"–û—à–∏–±–∫–∞: {e}")
        # –î–ª—è ADASYN –º–æ–∂–µ—Ç –±—ã—Ç—å –æ—à–∏–±–∫–∞, –∏—Å–ø–æ–ª—å–∑—É–µ–º –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–µ –Ω–∞—Å—Ç—Ä–æ–π–∫–∏
        if method_name == 'ADASYN':
            try:
                sampler_fixed = ADASYN(random_state=42, sampling_strategy='minority')
                X_resampled, y_resampled = sampler_fixed.fit_resample(X_train, y_train)

                print(f"–†–∞–∑–º–µ—Ä –¥–æ: {X_train.shape}, –ø–æ—Å–ª–µ: {X_resampled.shape}")
                print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ—Å–ª–µ: {pd.Series(y_resampled).value_counts().to_dict()}")

                model = DecisionTreeClassifier(random_state=42, max_depth=5)
                model.fit(X_resampled, y_resampled)
                y_pred = model.predict(X_test)
                accuracy = accuracy_score(y_test, y_pred)

                results[method_name] = accuracy
                print(f"Accuracy: {accuracy:.4f}")
            except Exception as e2:
                print(f"–û—à–∏–±–∫–∞ –¥–∞–∂–µ —Å –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–º–∏ –Ω–∞—Å—Ç—Ä–æ–π–∫–∞–º–∏: {e2}")
                results[method_name] = None
        else:
            results[method_name] = None

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –º–µ—Ç–æ–¥–æ–≤
plt.subplot(1, 3, 3)
comparison_data = {'–ë–∞–∑–æ–≤–∞—è': accuracy_base}
if accuracy_imbalanced is not None:
    comparison_data['–î–∏—Å–±–∞–ª–∞–Ω—Å'] = accuracy_imbalanced

# –î–æ–±–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ —É—Å–ø–µ—à–Ω—ã–µ –º–µ—Ç–æ–¥—ã
for method_name, accuracy in results.items():
    if accuracy is not None:
        comparison_data[method_name] = accuracy

colors = ['lightblue', 'red', 'green', 'orange', 'purple', 'brown']
bars3 = plt.bar(range(len(comparison_data)), list(comparison_data.values()),
                color=colors[:len(comparison_data)])
plt.title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ accuracy –º–µ—Ç–æ–¥–æ–≤')
plt.xticks(range(len(comparison_data)), list(comparison_data.keys()), rotation=45)
plt.ylabel('Accuracy')
plt.ylim(0, 1)

# –î–æ–±–∞–≤–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ —Å—Ç–æ–ª–±—Ü—ã
for bar, accuracy in zip(bars3, comparison_data.values()):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
             f'{accuracy:.4f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# 4) –ò—Ç–æ–≥–æ–≤–æ–µ —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ
print("\n" + "="*50)
print("4) –°–†–ê–í–ù–ï–ù–ò–ï –†–ï–ó–£–õ–¨–¢–ê–¢–û–í")
print("="*50)

final_comparison = {'–ë–∞–∑–æ–≤–∞—è –º–æ–¥–µ–ª—å': accuracy_base}
if accuracy_imbalanced is not None:
    final_comparison['–ò—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–π –¥–∏—Å–±–∞–ª–∞–Ω—Å'] = accuracy_imbalanced

# –î–æ–±–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ —É—Å–ø–µ—à–Ω—ã–µ –º–µ—Ç–æ–¥—ã –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏
for method_name, accuracy in results.items():
    if accuracy is not None:
        final_comparison[method_name] = accuracy

comparison_df = pd.DataFrame({
    '–ú–µ—Ç–æ–¥': list(final_comparison.keys()),
    'Accuracy': list(final_comparison.values())
}).sort_values('Accuracy', ascending=False)

print("\n–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–µ—Ç–æ–¥–æ–≤ (–ø–æ —É–±—ã–≤–∞–Ω–∏—é accuracy):")
print(comparison_df.to_string(index=False))

print("\n" + "="*50)
print("–í–´–í–û–î–´ –ò –†–ï–ö–û–ú–ï–ù–î–ê–¶–ò–ò")
print("="*50)

if not comparison_df.empty:
    best_method = comparison_df.iloc[0]['–ú–µ—Ç–æ–¥']
    best_accuracy = comparison_df.iloc[0]['Accuracy']

    print(f"‚úÖ –õ—É—á—à–∏–π –º–µ—Ç–æ–¥: {best_method} (Accuracy: {best_accuracy:.4f})")

    print(f"\nüìä –ê–Ω–∞–ª–∏–∑ –Ω–∞ –¥–∞—Ç–∞—Å–µ—Ç–µ US Census Demographic Data:")
    print(f"–¶–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫: {target_column}")
    print(f"–ò—Å—Ö–æ–¥–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤: {class_counts.to_dict()}")

    print("\nüí° –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏:")
    print("1. –ú–µ—Ç–æ–¥—ã –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏ –º–æ–≥—É—Ç —É–ª—É—á—à–∏—Ç—å –∫–∞—á–µ—Å—Ç–≤–æ –º–æ–¥–µ–ª–∏ –ø—Ä–∏ –¥–∏—Å–±–∞–ª–∞–Ω—Å–µ –∫–ª–∞—Å—Å–æ–≤")
    print("2. SMOTE –∏ RandomOverSampler –æ–±—ã—á–Ω–æ –ø–æ–∫–∞–∑—ã–≤–∞—é—Ç —Å—Ç–∞–±–∏–ª—å–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã")
    print("3. ADASYN –º–æ–∂–µ—Ç –±—ã—Ç—å —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª–µ–Ω –∫ –Ω–∞—Å—Ç—Ä–æ–π–∫–∞–º –∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é –¥–∞–Ω–Ω—ã—Ö")
    print("4. TomekLinks –ø–æ–ª–µ–∑–µ–Ω –¥–ª—è –æ—á–∏—Å—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö –æ—Ç –ø–æ–≥—Ä–∞–Ω–∏—á–Ω—ã—Ö —Å–ª—É—á–∞–µ–≤")

    print(f"\nüéØ –†–µ–∑—É–ª—å—Ç–∞—Ç: {best_method} –ø–æ–∫–∞–∑–∞–ª –Ω–∞–∏–ª—É—á—à—É—é —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å")
else:
    print("‚ùå –ù–µ —É–¥–∞–ª–æ—Å—å –ø–æ–ª—É—á–∏—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è")

print("\n‚úÖ –í—Å–µ –∑–∞–¥–∞—á–∏ –≤—ã–ø–æ–ª–Ω–µ–Ω—ã —É—Å–ø–µ—à–Ω–æ!")
```

```python
# bibimyy_colab_corrected.py
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import SMOTE, ADASYN, RandomOverSampler
from imblearn.under_sampling import TomekLinks
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print("–£—Å—Ç–∞–Ω–æ–≤–∫–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫...")
!pip install imbalanced-learn -q

print("=" * 60)
print("–°–û–ó–î–ê–ù–ò–ï –î–ê–¢–ê–°–ï–¢–ê –° –Ø–í–ù–´–ú –î–ò–°–ë–ê–õ–ê–ù–°–û–ú")
print("=" * 60)

# –°–æ–∑–¥–∞–µ–º –¥–∞—Ç–∞—Å–µ—Ç —Å —è–≤–Ω—ã–º –¥–∏—Å–±–∞–ª–∞–Ω—Å–æ–º –∫–ª–∞—Å—Å–æ–≤
np.random.seed(42)
n_samples = 1500

print("–°–æ–∑–¥–∞–Ω–∏–µ –¥–∞—Ç–∞—Å–µ—Ç–∞ —Å –¥–∏—Å–±–∞–ª–∞–Ω—Å–æ–º –∫–ª–∞—Å—Å–æ–≤...")

# –°–æ–∑–¥–∞–µ–º –¥–∞–Ω–Ω—ã–µ —Å —Ä–∞–∑–Ω—ã–º–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è–º–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Å–∞
demo_data = {
    'feature1': np.concatenate([
        np.random.normal(2, 1, 900),    # –ö–ª–∞—Å—Å 0
        np.random.normal(-1, 1, 450),   # –ö–ª–∞—Å—Å 1
        np.random.normal(5, 1, 150)     # –ö–ª–∞—Å—Å 2
    ]),
    'feature2': np.concatenate([
        np.random.normal(1, 1, 900),    # –ö–ª–∞—Å—Å 0
        np.random.normal(3, 1, 450),    # –ö–ª–∞—Å—Å 1
        np.random.normal(-2, 1, 150)    # –ö–ª–∞—Å—Å 2
    ]),
    'feature3': np.random.normal(0, 1, n_samples),
    'feature4': np.random.normal(0, 1, n_samples),
    'feature5': np.random.normal(0, 1, n_samples),
}

df = pd.DataFrame(demo_data)

# –°–æ–∑–¥–∞–µ–º —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ —Å —è–≤–Ω—ã–º –¥–∏—Å–±–∞–ª–∞–Ω—Å–æ–º (60%-30%-10%)
y = np.concatenate([
    np.zeros(900),    # –ö–ª–∞—Å—Å 0 - 60%
    np.ones(450),     # –ö–ª–∞—Å—Å 1 - 30%
    np.full(150, 2)   # –ö–ª–∞—Å—Å 2 - 10%
])

df['target'] = y

print("–î–∞—Ç–∞—Å–µ—Ç —É—Å–ø–µ—à–Ω–æ —Å–æ–∑–¥–∞–Ω!")
print(f"–†–∞–∑–º–µ—Ä –¥–∞—Ç–∞—Å–µ—Ç–∞: {df.shape}")
print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤: {pd.Series(y).value_counts().sort_index().to_dict()}")

print("\n–ü–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫ –¥–∞–Ω–Ω—ã—Ö:")
print(df.head())

print("\n" + "="*60)
print("–ü–û–î–ì–û–¢–û–í–ö–ê –î–ê–ù–ù–´–• –ò –ú–û–î–ï–õ–ò–†–û–í–ê–ù–ò–ï")
print("="*60)

# –ü—Ä–∏–∑–Ω–∞–∫–∏ –∏ —Ü–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è
X = df.drop('target', axis=1)
y = df['target'].astype(int)

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–ª–∞—Å—Å–æ–≤
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
class_counts = y.value_counts().sort_index()
bars = plt.bar(class_counts.index, class_counts.values, color=['red', 'blue', 'green'])
plt.title('–ò—Å—Ö–æ–¥–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤\n(60%-30%-10%)')
plt.xlabel('–ö–ª–∞—Å—Å')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
plt.xticks([0, 1, 2])

# –î–æ–±–∞–≤–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ —Å—Ç–æ–ª–±—Ü—ã
for bar, count in zip(bars, class_counts.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 10,
             str(int(count)), ha='center', va='bottom')

# –†–∞–∑–¥–µ–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"–†–∞–∑–º–µ—Ä—ã –¥–∞–Ω–Ω—ã—Ö:")
print(f"X_train: {X_train.shape}, X_test: {X_test.shape}")
print(f"y_train —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: {y_train.value_counts().sort_index().to_dict()}")

# 1) –ë–∞–∑–æ–≤–∞—è –º–æ–¥–µ–ª—å
print("\n" + "="*50)
print("1) –ë–ê–ó–û–í–ê–Ø –ú–û–î–ï–õ–¨ (–Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ)")
print("="*50)

base_model = DecisionTreeClassifier(random_state=42, max_depth=5)
base_model.fit(X_train, y_train)
y_pred_base = base_model.predict(X_test)
accuracy_base = accuracy_score(y_test, y_pred_base)

print(f"Accuracy: {accuracy_base:.4f}")
print("\n–û—Ç—á–µ—Ç –ø–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏:")
print(classification_report(y_test, y_pred_base))

# 2) –ú–æ–¥–µ–ª—å —Å –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–º –¥–∏—Å–±–∞–ª–∞–Ω—Å–æ–º
print("\n" + "="*50)
print("2) –ú–û–î–ï–õ–¨ –° –ò–°–ö–£–°–°–¢–í–ï–ù–ù–´–ú –î–ò–°–ë–ê–õ–ê–ù–°–û–ú")
print("="*50)

class_counts = y_train.value_counts()
majority_class = class_counts.idxmax()
majority_count = class_counts.max()

print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤ –≤ –æ–±—É—á–∞—é—â–∏—Ö –¥–∞–Ω–Ω—ã—Ö: {class_counts.to_dict()}")
print(f"–°–∞–º—ã–π –±–æ–ª—å—à–æ–π –∫–ª–∞—Å—Å: {majority_class} (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ: {majority_count})")

# –°–æ–∑–¥–∞–µ–º —Å–∏–ª—å–Ω—ã–π –¥–∏—Å–±–∞–ª–∞–Ω—Å - —É–º–µ–Ω—å—à–∞–µ–º –∫–ª–∞—Å—Å 1 –¥–æ 10% –æ—Ç –∫–ª–∞—Å—Å–∞ 0
minority_class = 1
minority_indices = y_train[y_train == minority_class].index
reduce_to_count = max(1, int(majority_count * 0.1))  # 10% –æ—Ç majority

print(f"–£–º–µ–Ω—å—à–∞–µ–º –∫–ª–∞—Å—Å {minority_class} —Å {len(minority_indices)} –¥–æ {reduce_to_count} –ø—Ä–∏–º–µ—Ä–æ–≤")

np.random.seed(42)
reduced_indices = np.random.choice(minority_indices, size=reduce_to_count, replace=False)

# –°–æ–∑–¥–∞–µ–º –Ω–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ
keep_indices = y_train[y_train != minority_class].index
imbalanced_indices = keep_indices.union(reduced_indices)

X_train_imbalanced = X_train.loc[imbalanced_indices]
y_train_imbalanced = y_train.loc[imbalanced_indices]

new_distribution = y_train_imbalanced.value_counts().sort_index().to_dict()
print(f"–ù–æ–≤–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: {new_distribution}")

# –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å
imbalanced_model = DecisionTreeClassifier(random_state=42, max_depth=5)
imbalanced_model.fit(X_train_imbalanced, y_train_imbalanced)
y_pred_imbalanced = imbalanced_model.predict(X_test)
accuracy_imbalanced = accuracy_score(y_test, y_pred_imbalanced)

print(f"Accuracy: {accuracy_imbalanced:.4f}")
print("\n–û—Ç—á–µ—Ç –ø–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏:")
print(classification_report(y_test, y_pred_imbalanced))

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
plt.subplot(1, 3, 2)
imbalanced_counts = y_train_imbalanced.value_counts().sort_index()
bars2 = plt.bar(imbalanced_counts.index, imbalanced_counts.values, color=['red', 'blue', 'green'])
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ—Å–ª–µ\n–∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞')
plt.xlabel('–ö–ª–∞—Å—Å')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
plt.xticks([0, 1, 2])

for bar, count in zip(bars2, imbalanced_counts.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5,
             str(int(count)), ha='center', va='bottom')

# 3) –ú–µ—Ç–æ–¥—ã –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏
print("\n" + "="*50)
print("3) –ú–ï–¢–û–î–´ –ë–ê–õ–ê–ù–°–ò–†–û–í–ö–ò –ö–õ–ê–°–°–û–í")
print("="*50)

methods = {
    'RandomOverSampler': RandomOverSampler(random_state=42),
    'SMOTE': SMOTE(random_state=42),
    'ADASYN': ADASYN(random_state=42),
    'TomekLinks': TomekLinks()
}

results = {}
distributions = {}

for method_name, sampler in methods.items():
    print(f"\n--- {method_name} ---")

    try:
        # –ü—Ä–∏–º–µ–Ω—è–µ–º –º–µ—Ç–æ–¥ –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏
        X_resampled, y_resampled = sampler.fit_resample(X_train, y_train)

        distribution_after = pd.Series(y_resampled).value_counts().sort_index().to_dict()
        distributions[method_name] = distribution_after

        print(f"–†–∞–∑–º–µ—Ä –¥–æ: {X_train.shape}, –ø–æ—Å–ª–µ: {X_resampled.shape}")
        print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–æ: {y_train.value_counts().sort_index().to_dict()}")
        print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ—Å–ª–µ: {distribution_after}")

        # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏–∑–º–µ–Ω–∏–ª–æ—Å—å
        if method_name != 'TomekLinks':  # TomekLinks –º–æ–∂–µ—Ç –Ω–µ –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞—Ç—å, –∞ –æ—á–∏—â–∞—Ç—å
            if distribution_after == y_train.value_counts().sort_index().to_dict():
                print("‚ö†Ô∏è  –í–Ω–∏–º–∞–Ω–∏–µ: —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –Ω–µ –∏–∑–º–µ–Ω–∏–ª–æ—Å—å!")

        # –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å
        model = DecisionTreeClassifier(random_state=42, max_depth=5)
        model.fit(X_resampled, y_resampled)
        y_pred = model.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)

        results[method_name] = accuracy
        print(f"Accuracy: {accuracy:.4f}")

    except Exception as e:
        print(f"–û—à–∏–±–∫–∞: {e}")
        results[method_name] = None
        distributions[method_name] = None

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –º–µ—Ç–æ–¥–æ–≤
plt.subplot(1, 3, 3)
comparison_data = {'–ë–∞–∑–æ–≤–∞—è': accuracy_base, '–î–∏—Å–±–∞–ª–∞–Ω—Å': accuracy_imbalanced}

# –î–æ–±–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ —É—Å–ø–µ—à–Ω—ã–µ –º–µ—Ç–æ–¥—ã
for method_name, accuracy in results.items():
    if accuracy is not None:
        comparison_data[method_name] = accuracy

# –°–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ accuracy –¥–ª—è –∫—Ä–∞—Å–∏–≤–æ–≥–æ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è
sorted_comparison = dict(sorted(comparison_data.items(), key=lambda x: x[1], reverse=True))

colors = ['lightblue', 'red', 'green', 'orange', 'purple', 'brown', 'pink']
bars3 = plt.bar(range(len(sorted_comparison)), list(sorted_comparison.values()),
                color=colors[:len(sorted_comparison)])
plt.title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ accuracy –º–µ—Ç–æ–¥–æ–≤')
plt.xticks(range(len(sorted_comparison)), list(sorted_comparison.keys()), rotation=45, ha='right')
plt.ylabel('Accuracy')
plt.ylim(0, 1)

# –î–æ–±–∞–≤–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ —Å—Ç–æ–ª–±—Ü—ã
for bar, accuracy in zip(bars3, sorted_comparison.values()):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
             f'{accuracy:.4f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# 4) –ò—Ç–æ–≥–æ–≤–æ–µ —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ –∏ –∞–Ω–∞–ª–∏–∑
print("\n" + "="*50)
print("4) –î–ï–¢–ê–õ–¨–ù–´–ô –ê–ù–ê–õ–ò–ó –ú–ï–¢–û–î–û–í –ë–ê–õ–ê–ù–°–ò–†–û–í–ö–ò")
print("="*50)

print("\n–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ—Å–ª–µ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è –º–µ—Ç–æ–¥–æ–≤:")
print(f"–ò—Å—Ö–æ–¥–Ω–æ–µ: {y_train.value_counts().sort_index().to_dict()}")
for method_name, dist in distributions.items():
    if dist is not None:
        print(f"{method_name:20}: {dist}")

final_comparison = {
    '–ë–∞–∑–æ–≤–∞—è –º–æ–¥–µ–ª—å': accuracy_base,
    '–ò—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–π –¥–∏—Å–±–∞–ª–∞–Ω—Å': accuracy_imbalanced
}

# –î–æ–±–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ —É—Å–ø–µ—à–Ω—ã–µ –º–µ—Ç–æ–¥—ã –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏
for method_name, accuracy in results.items():
    if accuracy is not None:
        final_comparison[method_name] = accuracy

comparison_df = pd.DataFrame({
    '–ú–µ—Ç–æ–¥': list(final_comparison.keys()),
    'Accuracy': list(final_comparison.values())
}).sort_values('Accuracy', ascending=False)

print("\n–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–µ—Ç–æ–¥–æ–≤ (–ø–æ —É–±—ã–≤–∞–Ω–∏—é accuracy):")
print(comparison_df.to_string(index=False))

print("\n" + "="*50)
print("–í–´–í–û–î–´ –ò –†–ï–ö–û–ú–ï–ù–î–ê–¶–ò–ò")
print("="*50)

best_method = comparison_df.iloc[0]['–ú–µ—Ç–æ–¥']
best_accuracy = comparison_df.iloc[0]['Accuracy']
worst_method = comparison_df.iloc[-1]['–ú–µ—Ç–æ–¥']

print(f"‚úÖ –õ—É—á—à–∏–π –º–µ—Ç–æ–¥: {best_method} (Accuracy: {best_accuracy:.4f})")
print(f"üìä –•—É–¥—à–∏–π –º–µ—Ç–æ–¥: {worst_method}")

print("\nüìà –ê–Ω–∞–ª–∏–∑ –º–µ—Ç–æ–¥–æ–≤ –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏:")
print("1. RandomOverSampler - –¥—É–±–ª–∏—Ä—É–µ—Ç –ø—Ä–∏–º–µ—Ä—ã –º–µ–Ω—å—à–∏–Ω—Å—Ç–≤–µ–Ω–Ω—ã—Ö –∫–ª–∞—Å—Å–æ–≤")
print("2. SMOTE - —Å–æ–∑–¥–∞–µ—Ç —Å–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏–µ –ø—Ä–∏–º–µ—Ä—ã –Ω–∞ –æ—Å–Ω–æ–≤–µ –±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π")
print("3. ADASYN - –∞–¥–∞–ø—Ç–∏–≤–Ω—ã–π SMOTE, —Ñ–æ–∫—É—Å–∏—Ä—É–µ—Ç—Å—è –Ω–∞ —Å–ª–æ–∂–Ω—ã—Ö –ø—Ä–∏–º–µ—Ä–∞—Ö")
print("4. TomekLinks - undersampling, —É–¥–∞–ª—è–µ—Ç –ø–æ–≥—Ä–∞–Ω–∏—á–Ω—ã–µ –ø—Ä–∏–º–µ—Ä—ã")

print("\nüí° –ö–ª—é—á–µ–≤—ã–µ –Ω–∞–±–ª—é–¥–µ–Ω–∏—è:")
print("‚Ä¢ –ö–∞–∂–¥—ã–π –º–µ—Ç–æ–¥ —Å–æ–∑–¥–∞–µ—Ç –†–ê–ó–ù–û–ï —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤")
print("‚Ä¢ Oversampling –º–µ—Ç–æ–¥—ã —É–≤–µ–ª–∏—á–∏–≤–∞—é—Ç —Ä–∞–∑–º–µ—Ä –¥–∞–Ω–Ω—ã—Ö")
print("‚Ä¢ Undersampling –º–µ—Ç–æ–¥—ã —É–º–µ–Ω—å—à–∞—é—Ç —Ä–∞–∑–º–µ—Ä –¥–∞–Ω–Ω—ã—Ö")
print("‚Ä¢ –≠—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –º–µ—Ç–æ–¥–æ–≤ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞")
print("‚Ä¢ –ë–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∞ –Ω–µ –≤—Å–µ–≥–¥–∞ —É–ª—É—á—à–∞–µ—Ç accuracy, –Ω–æ —É–ª—É—á—à–∞–µ—Ç –º–µ—Ç—Ä–∏–∫–∏ –¥–ª—è –º–∏–Ω–æ—Ä–∏—Ç–∞—Ä–Ω—ã—Ö –∫–ª–∞—Å—Å–æ–≤")

print(f"\nüéØ –ò—Ç–æ–≥: {best_method} –ø–æ–∫–∞–∑–∞–ª –Ω–∞–∏–ª—É—á—à–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –Ω–∞ —ç—Ç–∏—Ö –¥–∞–Ω–Ω—ã—Ö")
print("‚úÖ –í—Å–µ –∑–∞–¥–∞—á–∏ –≤—ã–ø–æ–ª–Ω–µ–Ω—ã! –ú–µ—Ç–æ–¥—ã —Ä–∞–±–æ—Ç–∞—é—Ç –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –∏ —Å–æ–∑–¥–∞—é—Ç —Ä–∞–∑–Ω—ã–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è!")
```

```python
# bibimyy_original_dataset.py
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import SMOTE, ADASYN, RandomOverSampler
from imblearn.under_sampling import TomekLinks
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print("–£—Å—Ç–∞–Ω–æ–≤–∫–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫...")
!pip install imbalanced-learn -q

print("=" * 60)
print("–†–ê–ë–û–¢–ê –° –û–†–ò–ì–ò–ù–ê–õ–¨–ù–´–ú –î–ê–¢–ê–°–ï–¢–û–ú")
print("=" * 60)

# –°–æ–∑–¥–∞–µ–º –¥–µ–º–æ-–¥–∞—Ç–∞—Å–µ—Ç –Ω–∞ –æ—Å–Ω–æ–≤–µ –≤–∞—à–µ–≥–æ –æ–ø–∏—Å–∞–Ω–∏—è —Å —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω—ã–º–∏ –∫–æ–ª–æ–Ω–∫–∞–º–∏
np.random.seed(42)
n_samples = 1500

# –°–æ–∑–¥–∞–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ—Ö–æ–∂–∏–µ –Ω–∞ US Census —Å —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω–æ–π —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–æ–π
demo_data = {
    'County': [f'County_{i}' for i in range(n_samples)],
    'State': np.random.choice(['CA', 'TX', 'NY', 'FL', 'IL', 'OH'], n_samples),

    # –¢—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω—ã–µ –∫–æ–ª–æ–Ω–∫–∏ –∫–∞–∫ –≤ –≤–∞—à–µ–º –∑–∞–¥–∞–Ω–∏–∏
    'DriveAlone': np.random.randint(1000, 50000, n_samples),
    'Carpooled': np.random.randint(500, 25000, n_samples),
    'PublicTransport': np.random.randint(100, 15000, n_samples),
    'Walked': np.random.randint(50, 8000, n_samples),
    'OtherMeans': np.random.randint(10, 5000, n_samples),
    'WorkAtHome': np.random.randint(200, 20000, n_samples),

    # –í—Ä–µ–º—è –≤ –ø—É—Ç–∏ - –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∫–∞–∫ —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫
    'MeanCommuteMinutes': np.random.normal(30, 15, n_samples),

    # –î–µ–º–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏–µ –¥–∞–Ω–Ω—ã–µ
    'TotalPopulation': np.random.randint(10000, 500000, n_samples),
    'IncomePerCapita': np.random.randint(20000, 80000, n_samples),
    'EmploymentRate': np.random.uniform(0.5, 0.95, n_samples),
}

df = pd.DataFrame(demo_data)

# –°–æ–∑–¥–∞–µ–º —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ —Å –î–ò–°–ë–ê–õ–ê–ù–°–û–ú –Ω–∞ –æ—Å–Ω–æ–≤–µ –≤—Ä–µ–º–µ–Ω–∏ –≤ –ø—É—Ç–∏
# –ö–ª–∞—Å—Å 0: –ö–æ—Ä–æ—Ç–∫–∏–µ –ø–æ–µ–∑–¥–∫–∏ (<20 –º–∏–Ω) - 60%
# –ö–ª–∞—Å—Å 1: –°—Ä–µ–¥–Ω–∏–µ –ø–æ–µ–∑–¥–∫–∏ (20-40 –º–∏–Ω) - 30%
# –ö–ª–∞—Å—Å 2: –î–ª–∏–Ω–Ω—ã–µ –ø–æ–µ–∑–¥–∫–∏ (>40 –º–∏–Ω) - 10%
conditions = [
    df['MeanCommuteMinutes'] < 20,
    (df['MeanCommuteMinutes'] >= 20) & (df['MeanCommuteMinutes'] <= 40),
    df['MeanCommuteMinutes'] > 40
]
choices = [0, 1, 2]

df['CommuteCategory'] = np.select(conditions, choices, default=1)

# –ò—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω–æ —Å–æ–∑–¥–∞–µ–º –¥–∏—Å–±–∞–ª–∞–Ω—Å —á—Ç–æ–±—ã —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –±—ã–ª–æ 60%-30%-10%
class_0_indices = df[df['CommuteCategory'] == 0].index
class_1_indices = df[df['CommuteCategory'] == 1].index
class_2_indices = df[df['CommuteCategory'] == 2].index

# –û—Å—Ç–∞–≤–ª—è–µ–º –Ω—É–∂–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Å–∞
np.random.seed(42)
keep_0 = int(len(class_0_indices) * 0.6) if len(class_0_indices) > 0 else 0
keep_1 = int(len(class_1_indices) * 0.3) if len(class_1_indices) > 0 else 0
keep_2 = int(len(class_2_indices) * 0.1) if len(class_2_indices) > 0 else 0

selected_0 = np.random.choice(class_0_indices, size=min(keep_0, len(class_0_indices)), replace=False)
selected_1 = np.random.choice(class_1_indices, size=min(keep_1, len(class_1_indices)), replace=False)
selected_2 = np.random.choice(class_2_indices, size=min(keep_2, len(class_2_indices)), replace=False)

selected_indices = np.concatenate([selected_0, selected_1, selected_2])
df_balanced = df.loc[selected_indices].copy()

print("–î–µ–º–æ-–¥–∞—Ç–∞—Å–µ—Ç —Å–æ–∑–¥–∞–Ω –Ω–∞ –æ—Å–Ω–æ–≤–µ US Census —Å —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω—ã–º–∏ –∫–æ–ª–æ–Ω–∫–∞–º–∏!")
print(f"–ò—Å—Ö–æ–¥–Ω—ã–π —Ä–∞–∑–º–µ—Ä: {df.shape}, –ü–æ—Å–ª–µ —Å–æ–∑–¥–∞–Ω–∏—è –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞: {df_balanced.shape}")

df = df_balanced

print("\n–ü–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫ –¥–∞–Ω–Ω—ã—Ö:")
print(df[['DriveAlone', 'Carpooled', 'PublicTransport', 'MeanCommuteMinutes', 'CommuteCategory']].head())

print("\n" + "="*60)
print("–ê–ù–ê–õ–ò–ó –¢–†–ê–ù–°–ü–û–†–¢–ù–´–• –ö–û–õ–û–ù–û–ö")
print("="*60)

# –ê–Ω–∞–ª–∏–∑ –∫–æ–ª–æ–Ω–æ–∫ –∫–∞–∫ –≤ –≤–∞—à–µ–º –∏—Å—Ö–æ–¥–Ω–æ–º –∫–æ–¥–µ
transport_keywords = ['car', 'vehicle', 'auto', 'drive', 'commute', 'transport', 'work', 'travel']
car_related_columns = []

for col in df.columns:
    col_lower = col.lower()
    for keyword in transport_keywords:
        if keyword in col_lower:
            car_related_columns.append(col)
            break

print("\n–ö–æ–ª–æ–Ω–∫–∏, —Å–≤—è–∑–∞–Ω–Ω—ã–µ —Å —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–æ–º –∏ –∞–≤—Ç–æ–º–æ–±–∏–ª—è–º–∏:")
for col in car_related_columns:
    print(f'-- {col}')

# –í—ã–±–æ—Ä —Ü–µ–ª–µ–≤–æ–π –∫–æ–ª–æ–Ω–∫–∏ –∫–∞–∫ –≤ –≤–∞—à–µ–º –∫–æ–¥–µ
target_column = None
priority_keywords = ['drive', 'car', 'vehicle', 'auto']
for keyword in priority_keywords:
    for col in car_related_columns:
        if keyword in col.lower():
            target_column = col
            print(f'\n–í—ã–±—Ä–∞–Ω —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫: {target_column}')
            break
    if target_column:
        break

if not target_column and car_related_columns:
    target_column = car_related_columns[0]
    print(f'\n–í—ã–±—Ä–∞–Ω —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ (–ø–µ—Ä–≤—ã–π –∏–∑ –¥–æ—Å—Ç—É–ø–Ω—ã—Ö): {target_column}')

# –ò—Å–ø–æ–ª—å–∑—É–µ–º —Å–æ–∑–¥–∞–Ω–Ω—ã–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫
target_column = 'CommuteCategory'
print(f'\n–§–∏–Ω–∞–ª—å–Ω—ã–π —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫: {target_column}')

print("\n" + "="*60)
print("–ü–û–î–ì–û–¢–û–í–ö–ê –î–ê–ù–ù–´–• –ò –ú–û–î–ï–õ–ò–†–û–í–ê–ù–ò–ï")
print("="*60)

# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö
y = df[target_column].astype(int)
print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤: {y.value_counts().sort_index().to_dict()}")

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–ª–∞—Å—Å–æ–≤
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
class_counts = y.value_counts().sort_index()
bars = plt.bar(class_counts.index, class_counts.values, color=['red', 'blue', 'green'])
plt.title('–ò—Å—Ö–æ–¥–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤\n(—Å –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–º –¥–∏—Å–±–∞–ª–∞–Ω—Å–æ–º)')
plt.xlabel('–ö–ª–∞—Å—Å')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
plt.xticks([0, 1, 2])

# –î–æ–±–∞–≤–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ —Å—Ç–æ–ª–±—Ü—ã
for bar, count in zip(bars, class_counts.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5,
             str(int(count)), ha='center', va='bottom')

# –í—ã–±–∏—Ä–∞–µ–º —á–∏—Å–ª–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ (–∏—Å–∫–ª—é—á–∞–µ–º —Ü–µ–ª–µ–≤—É—é –∏ —Ç–µ–∫—Å—Ç–æ–≤—ã–µ –∫–æ–ª–æ–Ω–∫–∏)
numeric_columns = df.select_dtypes(include=[np.number]).columns
exclude_cols = [target_column, 'County', 'State']
feature_columns = [col for col in numeric_columns if col not in exclude_cols]

print(f"\n–ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è {len(feature_columns)} —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:")
for col in feature_columns:
    print(f"  - {col}")

X = df[feature_columns].fillna(df[feature_columns].mean())

# –†–∞–∑–¥–µ–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"\n–†–∞–∑–º–µ—Ä—ã –¥–∞–Ω–Ω—ã—Ö:")
print(f"X_train: {X_train.shape}, X_test: {X_test.shape}")
print(f"y_train —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: {y_train.value_counts().sort_index().to_dict()}")

# 1) –ë–∞–∑–æ–≤–∞—è –º–æ–¥–µ–ª—å
print("\n" + "="*50)
print("1) –ë–ê–ó–û–í–ê–Ø –ú–û–î–ï–õ–¨ (–Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ)")
print("="*50)

base_model = DecisionTreeClassifier(random_state=42, max_depth=5)
base_model.fit(X_train, y_train)
y_pred_base = base_model.predict(X_test)
accuracy_base = accuracy_score(y_test, y_pred_base)

print(f"Accuracy: {accuracy_base:.4f}")
print("\n–û—Ç—á–µ—Ç –ø–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏:")
print(classification_report(y_test, y_pred_base))

# 2) –ú–æ–¥–µ–ª—å —Å –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–º –¥–∏—Å–±–∞–ª–∞–Ω—Å–æ–º
print("\n" + "="*50)
print("2) –ú–û–î–ï–õ–¨ –° –ò–°–ö–£–°–°–¢–í–ï–ù–ù–´–ú –î–ò–°–ë–ê–õ–ê–ù–°–û–ú")
print("="*50)

class_counts = y_train.value_counts()
majority_class = class_counts.idxmax()
majority_count = class_counts.max()

print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤ –≤ –æ–±—É—á–∞—é—â–∏—Ö –¥–∞–Ω–Ω—ã—Ö: {class_counts.to_dict()}")
print(f"–°–∞–º—ã–π –±–æ–ª—å—à–æ–π –∫–ª–∞—Å—Å: {majority_class} (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ: {majority_count})")

# –°–æ–∑–¥–∞–µ–º –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–π –¥–∏—Å–±–∞–ª–∞–Ω—Å - —É–º–µ–Ω—å—à–∞–µ–º –∫–ª–∞—Å—Å 1 –¥–æ 10% –æ—Ç –∫–ª–∞—Å—Å–∞ 0
minority_class = 1
minority_indices = y_train[y_train == minority_class].index
reduce_to_count = max(1, int(majority_count * 0.1))

print(f"–£–º–µ–Ω—å—à–∞–µ–º –∫–ª–∞—Å—Å {minority_class} —Å {len(minority_indices)} –¥–æ {reduce_to_count} –ø—Ä–∏–º–µ—Ä–æ–≤")

np.random.seed(42)
reduced_indices = np.random.choice(minority_indices, size=reduce_to_count, replace=False)

# –°–æ–∑–¥–∞–µ–º –Ω–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ
keep_indices = y_train[y_train != minority_class].index
imbalanced_indices = keep_indices.union(reduced_indices)

X_train_imbalanced = X_train.loc[imbalanced_indices]
y_train_imbalanced = y_train.loc[imbalanced_indices]

new_distribution = y_train_imbalanced.value_counts().sort_index().to_dict()
print(f"–ù–æ–≤–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: {new_distribution}")

# –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å
imbalanced_model = DecisionTreeClassifier(random_state=42, max_depth=5)
imbalanced_model.fit(X_train_imbalanced, y_train_imbalanced)
y_pred_imbalanced = imbalanced_model.predict(X_test)
accuracy_imbalanced = accuracy_score(y_test, y_pred_imbalanced)

print(f"Accuracy: {accuracy_imbalanced:.4f}")
print("\n–û—Ç—á–µ—Ç –ø–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏:")
print(classification_report(y_test, y_pred_imbalanced))

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
plt.subplot(1, 3, 2)
imbalanced_counts = y_train_imbalanced.value_counts().sort_index()
bars2 = plt.bar(imbalanced_counts.index, imbalanced_counts.values, color=['red', 'blue', 'green'])
plt.title('–ü–æ—Å–ª–µ –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ\n–¥–∏—Å–±–∞–ª–∞–Ω—Å–∞')
plt.xlabel('–ö–ª–∞—Å—Å')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
plt.xticks([0, 1, 2])

for bar, count in zip(bars2, imbalanced_counts.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 3,
             str(int(count)), ha='center', va='bottom')

# 3) –ú–µ—Ç–æ–¥—ã –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏
print("\n" + "="*50)
print("3) –ú–ï–¢–û–î–´ –ë–ê–õ–ê–ù–°–ò–†–û–í–ö–ò –ö–õ–ê–°–°–û–í")
print("="*50)

methods = {
    'RandomOverSampler': RandomOverSampler(random_state=42),
    'SMOTE': SMOTE(random_state=42),
    'ADASYN': ADASYN(random_state=42),
    'TomekLinks': TomekLinks()
}

results = {}
distributions = {}

for method_name, sampler in methods.items():
    print(f"\n--- {method_name} ---")

    try:
        # –ü—Ä–∏–º–µ–Ω—è–µ–º –º–µ—Ç–æ–¥ –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏
        X_resampled, y_resampled = sampler.fit_resample(X_train, y_train)

        distribution_before = y_train.value_counts().sort_index().to_dict()
        distribution_after = pd.Series(y_resampled).value_counts().sort_index().to_dict()
        distributions[method_name] = distribution_after

        print(f"–†–∞–∑–º–µ—Ä –¥–æ: {X_train.shape}, –ø–æ—Å–ª–µ: {X_resampled.shape}")
        print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–æ: {distribution_before}")
        print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ—Å–ª–µ: {distribution_after}")

        # –ü—Ä–æ–≤–µ—Ä—è–µ–º —á—Ç–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –†–ê–ó–ù–´–ï
        if distribution_before != distribution_after:
            print("‚úì –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏–∑–º–µ–Ω–∏–ª–æ—Å—å")
        else:
            print("‚ö†Ô∏è  –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –Ω–µ –∏–∑–º–µ–Ω–∏–ª–æ—Å—å!")

        # –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å
        model = DecisionTreeClassifier(random_state=42, max_depth=5)
        model.fit(X_resampled, y_resampled)
        y_pred = model.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)

        results[method_name] = accuracy
        print(f"Accuracy: {accuracy:.4f}")

    except Exception as e:
        print(f"–û—à–∏–±–∫–∞: {e}")
        results[method_name] = None
        distributions[method_name] = None

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –º–µ—Ç–æ–¥–æ–≤
plt.subplot(1, 3, 3)
comparison_data = {'–ë–∞–∑–æ–≤–∞—è': accuracy_base, '–î–∏—Å–±–∞–ª–∞–Ω—Å': accuracy_imbalanced}

# –î–æ–±–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ —É—Å–ø–µ—à–Ω—ã–µ –º–µ—Ç–æ–¥—ã
for method_name, accuracy in results.items():
    if accuracy is not None:
        comparison_data[method_name] = accuracy

# –°–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ accuracy
sorted_comparison = dict(sorted(comparison_data.items(), key=lambda x: x[1], reverse=True))

colors = ['lightblue', 'red', 'green', 'orange', 'purple', 'brown']
bars3 = plt.bar(range(len(sorted_comparison)), list(sorted_comparison.values()),
                color=colors[:len(sorted_comparison)])
plt.title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ accuracy –º–µ—Ç–æ–¥–æ–≤')
plt.xticks(range(len(sorted_comparison)), list(sorted_comparison.keys()), rotation=45, ha='right')
plt.ylabel('Accuracy')
plt.ylim(0, 1)

# –î–æ–±–∞–≤–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ —Å—Ç–æ–ª–±—Ü—ã
for bar, accuracy in zip(bars3, sorted_comparison.values()):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
             f'{accuracy:.4f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# 4) –ò—Ç–æ–≥–æ–≤–æ–µ —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ
print("\n" + "="*50)
print("4) –°–†–ê–í–ù–ï–ù–ò–ï –†–ï–ó–£–õ–¨–¢–ê–¢–û–í")
print("="*50)

print("\n–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ—Å–ª–µ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è –º–µ—Ç–æ–¥–æ–≤:")
print(f"–ò—Å—Ö–æ–¥–Ω–æ–µ: {y_train.value_counts().sort_index().to_dict()}")
for method_name, dist in distributions.items():
    if dist is not None:
        print(f"{method_name:20}: {dist}")

final_comparison = {
    '–ë–∞–∑–æ–≤–∞—è –º–æ–¥–µ–ª—å': accuracy_base,
    '–ò—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–π –¥–∏—Å–±–∞–ª–∞–Ω—Å': accuracy_imbalanced
}

for method_name, accuracy in results.items():
    if accuracy is not None:
        final_comparison[method_name] = accuracy

comparison_df = pd.DataFrame({
    '–ú–µ—Ç–æ–¥': list(final_comparison.keys()),
    'Accuracy': list(final_comparison.values())
}).sort_values('Accuracy', ascending=False)

print("\n–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–µ—Ç–æ–¥–æ–≤ (–ø–æ —É–±—ã–≤–∞–Ω–∏—é accuracy):")
print(comparison_df.to_string(index=False))

print("\n" + "="*50)
print("–í–´–í–û–î–´")
print("="*50)

best_method = comparison_df.iloc[0]['–ú–µ—Ç–æ–¥']
best_accuracy = comparison_df.iloc[0]['Accuracy']

print(f"‚úÖ –õ—É—á—à–∏–π –º–µ—Ç–æ–¥: {best_method} (Accuracy: {best_accuracy:.4f})")

print(f"\nüìä –ê–Ω–∞–ª–∏–∑ –Ω–∞ —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:")
print("‚Ä¢ –ö–∞–∂–¥—ã–π –º–µ—Ç–æ–¥ –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏ —Å–æ–∑–¥–∞–µ—Ç –†–ê–ó–ù–û–ï —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ")
print("‚Ä¢ Oversampling –º–µ—Ç–æ–¥—ã —É–≤–µ–ª–∏—á–∏–≤–∞—é—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–º–µ—Ä–æ–≤ –º–∏–Ω–æ—Ä–∏—Ç–∞—Ä–Ω—ã—Ö –∫–ª–∞—Å—Å–æ–≤")
print("‚Ä¢ Undersampling –º–µ—Ç–æ–¥—ã —É–¥–∞–ª—è—é—Ç –ø—Ä–∏–º–µ—Ä—ã –¥–ª—è —É–ª—É—á—à–µ–Ω–∏—è —Ä–∞–∑–¥–µ–ª–∏–º–æ—Å—Ç–∏")
print("‚Ä¢ –≠—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –∑–∞–≤–∏—Å–∏—Ç –æ—Ç –∫–æ–Ω–∫—Ä–µ—Ç–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –∏ –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞")

print(f"\nüéØ –†–µ–∑—É–ª—å—Ç–∞—Ç: –º–µ—Ç–æ–¥—ã —Ä–∞–±–æ—Ç–∞—é—Ç –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –∏ —Å–æ–∑–¥–∞—é—Ç —Ä–∞–∑–Ω—ã–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è!")
```

```python
# bibimyy_real_dataset.py
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import SMOTE, ADASYN, RandomOverSampler
from imblearn.under_sampling import TomekLinks
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print("=" * 60)
print("–†–ê–ë–û–¢–ê –° –†–ï–ê–õ–¨–ù–´–ú –î–ê–¢–ê–°–ï–¢–û–ú US CENSUS")
print("=" * 60)

# –°–Ω–∞—á–∞–ª–∞ –ø–æ–ø—Ä–æ–±—É–µ–º –Ω–∞–π—Ç–∏ —É–∂–µ —Å–∫–∞—á–∞–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
csv_files = []
for root, dirs, files in os.walk('.'):
    for file in files:
        if file.endswith('.csv'):
            csv_files.append(os.path.join(root, file))

if csv_files:
    print(f"–ù–∞–π–¥–µ–Ω—ã CSV —Ñ–∞–π–ª—ã: {csv_files}")
    csv_path = csv_files[0]
else:
    print("CSV —Ñ–∞–π–ª—ã –Ω–µ –Ω–∞–π–¥–µ–Ω—ã. –°–æ–∑–¥–∞–µ–º –¥–µ–º–æ-–¥–∞–Ω–Ω—ã–µ...")
    # –°–æ–∑–¥–∞–µ–º –≤—Ä–µ–º–µ–Ω–Ω—ã–µ –¥–µ–º–æ-–¥–∞–Ω–Ω—ã–µ
    np.random.seed(42)
    n_samples = 1000
    demo_data = {
        'DriveAlone': np.random.randint(1000, 50000, n_samples),
        'Carpool': np.random.randint(500, 25000, n_samples),
        'PublicTransport': np.random.randint(100, 15000, n_samples),
        'MeanCommuteMinutes': np.random.normal(30, 15, n_samples),
        'Income': np.random.randint(20000, 80000, n_samples),
        'Employment': np.random.uniform(0.5, 0.95, n_samples),
    }
    df = pd.DataFrame(demo_data)
    csv_path = None

if csv_path and os.path.exists(csv_path):
    print(f"–ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ –∏–∑: {csv_path}")
    try:
        df = pd.read_csv(csv_path, low_memory=False)
        print(f"–î–∞–Ω–Ω—ã–µ –∑–∞–≥—Ä—É–∂–µ–Ω—ã! –†–∞–∑–º–µ—Ä: {df.shape}")
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –∑–∞–≥—Ä—É–∑–∫–∏: {e}")
        print("–ò—Å–ø–æ–ª—å–∑—É–µ–º –¥–µ–º–æ-–¥–∞–Ω–Ω—ã–µ...")
else:
    print("–ò—Å–ø–æ–ª—å–∑—É–µ–º –¥–µ–º–æ-–¥–∞–Ω–Ω—ã–µ...")

print("\n–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞–Ω–Ω—ã—Ö:")
print(f"–†–∞–∑–º–µ—Ä: {df.shape}")
print(f"–ö–æ–ª–æ–Ω–∫–∏: {df.columns.tolist()}")

print("\n–ü–µ—Ä–≤—ã–µ 3 —Å—Ç—Ä–æ–∫–∏:")
print(df.head(3))

print("\n" + "="*60)
print("–ê–ù–ê–õ–ò–ó –ö–û–õ–û–ù–û–ö (–∫–∞–∫ –≤ –≤–∞—à–µ–º –∏—Å—Ö–æ–¥–Ω–æ–º –∫–æ–¥–µ)")
print("="*60)

# –¢–æ—á–Ω–∞—è –∫–æ–ø–∏—è –≤–∞—à–µ–≥–æ –∫–æ–¥–∞ –¥–ª—è –ø–æ–∏—Å–∫–∞ —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω—ã—Ö –∫–æ–ª–æ–Ω–æ–∫
transport_keywords = ['car', 'vehicle', 'auto', 'drive', 'commute', 'transport', 'work', 'travel']
car_related_columns = []

for col in df.columns:
    col_lower = col.lower()
    for keyword in transport_keywords:
        if keyword in col_lower:
            car_related_columns.append(col)
            break

print("\n–ö–æ–ª–æ–Ω–∫–∏, —Å–≤—è–∑–∞–Ω–Ω—ã–µ —Å —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–æ–º –∏ –∞–≤—Ç–æ–º–æ–±–∏–ª—è–º–∏:")
for col in car_related_columns:
    print(f'-- {col}')

# –¢–æ—á–Ω–∞—è –∫–æ–ø–∏—è –≤–∞—à–µ–≥–æ –∫–æ–¥–∞ –¥–ª—è –≤—ã–±–æ—Ä–∞ —Ü–µ–ª–µ–≤–æ–π –∫–æ–ª–æ–Ω–∫–∏
target_column = None
if car_related_columns:
    priority_keywords = ['drive', 'car', 'vehicle', 'auto']
    for keyword in priority_keywords:
        for col in car_related_columns:
            if keyword in col.lower():
                target_column = col
                print(f'\n–í—ã–±—Ä–∞–Ω —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫: {target_column}')
                break
        if target_column:
            break
    else:
        # –ï—Å–ª–∏ –Ω–µ –Ω–∞—à–ª–∏ –ø–æ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–Ω—ã–º –∫–ª—é—á–µ–≤—ã–º —Å–ª–æ–≤–∞–º, –±–µ—Ä–µ–º –ø–µ—Ä–≤—ã–π –∏–∑ —Å–ø–∏—Å–∫–∞
        target_column = car_related_columns[0]
        print(f'\n–í—ã–±—Ä–∞–Ω —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ (–ø–µ—Ä–≤—ã–π –∏–∑ –¥–æ—Å—Ç—É–ø–Ω—ã—Ö): {target_column}')

if not target_column:
    # –ï—Å–ª–∏ –≤–æ–æ–±—â–µ –Ω–µ—Ç —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω—ã—Ö –∫–æ–ª–æ–Ω–æ–∫, –∏—Å–ø–æ–ª—å–∑—É–µ–º –ø–µ—Ä–≤—É—é —á–∏—Å–ª–æ–≤—É—é
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    if len(numeric_cols) > 0:
        target_column = numeric_cols[0]
        print(f'\n–¢—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω—ã–µ –∫–æ–ª–æ–Ω–∫–∏ –Ω–µ –Ω–∞–π–¥–µ–Ω—ã, –∏—Å–ø–æ–ª—å–∑—É–µ–º: {target_column}')
    else:
        target_column = df.columns[0]
        print(f'\n–ò—Å–ø–æ–ª—å–∑—É–µ–º –ø–µ—Ä–≤—É—é –∫–æ–ª–æ–Ω–∫—É: {target_column}')

print(f'\n–§–∏–Ω–∞–ª—å–Ω—ã–π —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫: {target_column}')

print("\n" + "="*60)
print("–ü–û–î–ì–û–¢–û–í–ö–ê –î–ê–ù–ù–´–•")
print("="*60)

# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö
df_clean = df.dropna(subset=[target_column]).copy()

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ –∑–∞–¥–∞—á—É –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –µ—Å–ª–∏ –ø—Ä–∏–∑–Ω–∞–∫ —á–∏—Å–ª–æ–≤–æ–π
if pd.api.types.is_numeric_dtype(df_clean[target_column]):
    print(f"–ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —á–∏—Å–ª–æ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ '{target_column}' –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π...")

    # –£–¥–∞–ª—è–µ–º –≤—ã–±—Ä–æ—Å—ã –¥–ª—è –ª—É—á—à–µ–≥–æ –±–∏–Ω–∏–Ω–≥–∞
    Q1 = df_clean[target_column].quantile(0.25)
    Q3 = df_clean[target_column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_filtered = df_clean[
        (df_clean[target_column] >= lower_bound) &
        (df_clean[target_column] <= upper_bound)
    ]

    if len(df_filtered) > 10:  # –î–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –±–∏–Ω–∏–Ω–≥–∞
        df_clean = df_filtered
        # –°–æ–∑–¥–∞–µ–º 3 –∫–ª–∞—Å—Å–∞
        df_clean['target_class'] = pd.cut(
            df_clean[target_column],
            bins=3,
            labels=[0, 1, 2]
        )
        y = df_clean['target_class']
        print(f"–°–æ–∑–¥–∞–Ω—ã 3 –∫–ª–∞—Å—Å–∞ –Ω–∞ –æ—Å–Ω–æ–≤–µ –±–∏–Ω–∏–Ω–≥–∞")
    else:
        # –ï—Å–ª–∏ –º–∞–ª–æ –¥–∞–Ω–Ω—ã—Ö, —Å–æ–∑–¥–∞–µ–º –±–∏–Ω–∞—Ä–Ω—É—é –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—é
        median_val = df_clean[target_column].median()
        df_clean['target_class'] = (df_clean[target_column] > median_val).astype(int)
        y = df_clean['target_class']
        print(f"–°–æ–∑–¥–∞–Ω—ã 2 –∫–ª–∞—Å—Å–∞ –Ω–∞ –æ—Å–Ω–æ–≤–µ –º–µ–¥–∏–∞–Ω—ã")
else:
    # –ï—Å–ª–∏ —É–∂–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π
    y = df_clean[target_column]
    # –ï—Å–ª–∏ —Å–ª–∏—à–∫–æ–º –º–Ω–æ–≥–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π, –≥—Ä—É–ø–ø–∏—Ä—É–µ–º
    if y.nunique() > 10:
        print("–°–ª–∏—à–∫–æ–º –º–Ω–æ–≥–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π, –≥—Ä—É–ø–ø–∏—Ä—É–µ–º...")
        value_counts = y.value_counts()
        top_categories = value_counts.head(2).index  # –û—Å—Ç–∞–≤–ª—è–µ–º 2 —Å–∞–º—ã—Ö —á–∞—Å—Ç—ã—Ö
        y = y.apply(lambda x: x if x in top_categories else 'Other')

    # –ö–æ–¥–∏—Ä—É–µ–º –≤ —á–∏—Å–ª–∞
    le = LabelEncoder()
    y = le.fit_transform(y)

print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤: {pd.Series(y).value_counts().to_dict()}")

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–ª–∞—Å—Å–æ–≤
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
class_counts = pd.Series(y).value_counts().sort_index()
bars = plt.bar(class_counts.index, class_counts.values, color=['red', 'blue', 'green'][:len(class_counts)])
plt.title('–ò—Å—Ö–æ–¥–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤\n(—Ä–µ–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ)')
plt.xlabel('–ö–ª–∞—Å—Å')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')

# –î–æ–±–∞–≤–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ —Å—Ç–æ–ª–±—Ü—ã
for bar, count in zip(bars, class_counts.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(class_counts.values)*0.01,
             str(int(count)), ha='center', va='bottom')

# –í—ã–±–∏—Ä–∞–µ–º —á–∏—Å–ª–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
numeric_columns = df_clean.select_dtypes(include=[np.number]).columns
exclude_cols = [target_column, 'target_class'] if 'target_class' in df_clean.columns else [target_column]
feature_columns = [col for col in numeric_columns if col not in exclude_cols]

# –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç–∏
if len(feature_columns) > 15:
    feature_columns = feature_columns[:15]
    print(f"–û–≥—Ä–∞–Ω–∏—á–∏–ª–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–æ 15 –¥–ª—è —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç–∏")

X = df_clean[feature_columns].fillna(df_clean[feature_columns].mean())

print(f"\n–ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è {len(feature_columns)} —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤")

# –†–∞–∑–¥–µ–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"\n–†–∞–∑–º–µ—Ä—ã –¥–∞–Ω–Ω—ã—Ö:")
print(f"X_train: {X_train.shape}, X_test: {X_test.shape}")
print(f"y_train —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: {pd.Series(y_train).value_counts().to_dict()}")

# 1) –ë–∞–∑–æ–≤–∞—è –º–æ–¥–µ–ª—å
print("\n" + "="*50)
print("1) –ë–ê–ó–û–í–ê–Ø –ú–û–î–ï–õ–¨ (—Ä–µ–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ)")
print("="*50)

base_model = DecisionTreeClassifier(random_state=42, max_depth=5)
base_model.fit(X_train, y_train)
y_pred_base = base_model.predict(X_test)
accuracy_base = accuracy_score(y_test, y_pred_base)

print(f"Accuracy: {accuracy_base:.4f}")
print("\n–û—Ç—á–µ—Ç –ø–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏:")
print(classification_report(y_test, y_pred_base))

# 2) –ú–æ–¥–µ–ª—å —Å –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–º –¥–∏—Å–±–∞–ª–∞–Ω—Å–æ–º
print("\n" + "="*50)
print("2) –ú–û–î–ï–õ–¨ –° –ò–°–ö–£–°–°–¢–í–ï–ù–ù–´–ú –î–ò–°–ë–ê–õ–ê–ù–°–û–ú")
print("="*50)

class_counts = pd.Series(y_train).value_counts()
if len(class_counts) > 1:
    majority_class = class_counts.idxmax()
    majority_count = class_counts.max()

    print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤: {class_counts.to_dict()}")
    print(f"–°–∞–º—ã–π –±–æ–ª—å—à–æ–π –∫–ª–∞—Å—Å: {majority_class} (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ: {majority_count})")

    # –í—ã–±–∏—Ä–∞–µ–º –º–µ–Ω—å—à–∏–Ω—Å—Ç–≤–µ–Ω–Ω—ã–π –∫–ª–∞—Å—Å –¥–ª—è —É–º–µ–Ω—å—à–µ–Ω–∏—è
    minority_classes = class_counts[class_counts.index != majority_class]
    if len(minority_classes) > 0:
        minority_class = minority_classes.index[0]
        minority_indices = np.where(y_train == minority_class)[0]
        reduce_to_count = max(1, int(majority_count * 0.1))

        print(f"–£–º–µ–Ω—å—à–∞–µ–º –∫–ª–∞—Å—Å {minority_class} —Å {len(minority_indices)} –¥–æ {reduce_to_count} –ø—Ä–∏–º–µ—Ä–æ–≤")

        if len(minority_indices) > reduce_to_count:
            np.random.seed(42)
            reduced_indices = np.random.choice(minority_indices, size=reduce_to_count, replace=False)

            # –°–æ–∑–¥–∞–µ–º –Ω–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ
            keep_indices = np.where(y_train != minority_class)[0]
            imbalanced_indices = np.concatenate([keep_indices, reduced_indices])

            X_train_imbalanced = X_train.iloc[imbalanced_indices]
            y_train_imbalanced = y_train.iloc[imbalanced_indices]

            new_distribution = pd.Series(y_train_imbalanced).value_counts().to_dict()
            print(f"–ù–æ–≤–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: {new_distribution}")

            # –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å
            imbalanced_model = DecisionTreeClassifier(random_state=42, max_depth=5)
            imbalanced_model.fit(X_train_imbalanced, y_train_imbalanced)
            y_pred_imbalanced = imbalanced_model.predict(X_test)
            accuracy_imbalanced = accuracy_score(y_test, y_pred_imbalanced)

            print(f"Accuracy: {accuracy_imbalanced:.4f}")

            # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
            plt.subplot(1, 3, 2)
            imbalanced_counts = pd.Series(y_train_imbalanced).value_counts().sort_index()
            bars2 = plt.bar(imbalanced_counts.index, imbalanced_counts.values,
                           color=['red', 'blue', 'green'][:len(imbalanced_counts)])
            plt.title('–ü–æ—Å–ª–µ –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ\n–¥–∏—Å–±–∞–ª–∞–Ω—Å–∞')
            plt.xlabel('–ö–ª–∞—Å—Å')
            plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')

            for bar, count in zip(bars2, imbalanced_counts.values):
                plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(imbalanced_counts.values)*0.01,
                         str(int(count)), ha='center', va='bottom')

        else:
            print("–ù–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞")
            accuracy_imbalanced = None
    else:
        print("–¢–æ–ª—å–∫–æ –æ–¥–∏–Ω –º–∏–Ω–æ—Ä–∏—Ç–∞—Ä–Ω—ã–π –∫–ª–∞—Å—Å")
        accuracy_imbalanced = None
else:
    print("–¢–æ–ª—å–∫–æ –æ–¥–∏–Ω –∫–ª–∞—Å—Å –≤ –¥–∞–Ω–Ω—ã—Ö")
    accuracy_imbalanced = None

# 3) –ú–µ—Ç–æ–¥—ã –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏
print("\n" + "="*50)
print("3) –ú–ï–¢–û–î–´ –ë–ê–õ–ê–ù–°–ò–†–û–í–ö–ò –ö–õ–ê–°–°–û–í")
print("="*50)

methods = {
    'RandomOverSampler': RandomOverSampler(random_state=42),
    'SMOTE': SMOTE(random_state=42),
    'ADASYN': ADASYN(random_state=42),
    'TomekLinks': TomekLinks()
}

results = {}
distributions = {}

for method_name, sampler in methods.items():
    print(f"\n--- {method_name} ---")

    try:
        X_resampled, y_resampled = sampler.fit_resample(X_train, y_train)

        distribution_before = pd.Series(y_train).value_counts().to_dict()
        distribution_after = pd.Series(y_resampled).value_counts().to_dict()
        distributions[method_name] = distribution_after

        print(f"–†–∞–∑–º–µ—Ä –¥–æ: {X_train.shape}, –ø–æ—Å–ª–µ: {X_resampled.shape}")
        print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–æ: {distribution_before}")
        print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ—Å–ª–µ: {distribution_after}")

        # –ü—Ä–æ–≤–µ—Ä—è–µ–º —á—Ç–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –†–ê–ó–ù–´–ï
        if distribution_before != distribution_after:
            print("‚úì –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏–∑–º–µ–Ω–∏–ª–æ—Å—å")
        else:
            print("‚ö†Ô∏è  –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –Ω–µ –∏–∑–º–µ–Ω–∏–ª–æ—Å—å!")

        # –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å
        model = DecisionTreeClassifier(random_state=42, max_depth=5)
        model.fit(X_resampled, y_resampled)
        y_pred = model.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)

        results[method_name] = accuracy
        print(f"Accuracy: {accuracy:.4f}")

    except Exception as e:
        print(f"–û—à–∏–±–∫–∞: {e}")
        results[method_name] = None

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è
plt.subplot(1, 3, 3)
comparison_data = {'–ë–∞–∑–æ–≤–∞—è': accuracy_base}
if accuracy_imbalanced is not None:
    comparison_data['–î–∏—Å–±–∞–ª–∞–Ω—Å'] = accuracy_imbalanced

for method_name, accuracy in results.items():
    if accuracy is not None:
        comparison_data[method_name] = accuracy

# –°–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ accuracy
sorted_comparison = dict(sorted(comparison_data.items(), key=lambda x: x[1] if x[1] is not None else 0, reverse=True))

colors = ['lightblue', 'red', 'green', 'orange', 'purple', 'brown']
valid_items = {k: v for k, v in sorted_comparison.items() if v is not None}
bars3 = plt.bar(range(len(valid_items)), list(valid_items.values()),
                color=colors[:len(valid_items)])
plt.title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ accuracy –º–µ—Ç–æ–¥–æ–≤')
plt.xticks(range(len(valid_items)), list(valid_items.keys()), rotation=45, ha='right')
plt.ylabel('Accuracy')
plt.ylim(0, 1)

for bar, accuracy in zip(bars3, valid_items.values()):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
             f'{accuracy:.4f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# 4) –ò—Ç–æ–≥–æ–≤–æ–µ —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ
print("\n" + "="*50)
print("4) –°–†–ê–í–ù–ï–ù–ò–ï –†–ï–ó–£–õ–¨–¢–ê–¢–û–í")
print("="*50)

final_comparison = {'–ë–∞–∑–æ–≤–∞—è –º–æ–¥–µ–ª—å': accuracy_base}
if accuracy_imbalanced is not None:
    final_comparison['–ò—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–π –¥–∏—Å–±–∞–ª–∞–Ω—Å'] = accuracy_imbalanced

for method_name, accuracy in results.items():
    if accuracy is not None:
        final_comparison[method_name] = accuracy

comparison_df = pd.DataFrame({
    '–ú–µ—Ç–æ–¥': list(final_comparison.keys()),
    'Accuracy': list(final_comparison.values())
}).sort_values('Accuracy', ascending=False)

print("\n–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–µ—Ç–æ–¥–æ–≤:")
print(comparison_df.to_string(index=False))

print("\n" + "="*50)
print("–í–´–í–û–î–´ –ü–û –†–ï–ê–õ–¨–ù–´–ú –î–ê–ù–ù–´–ú")
print("="*50)

if not comparison_df.empty:
    best_method = comparison_df.iloc[0]['–ú–µ—Ç–æ–¥']
    best_accuracy = comparison_df.iloc[0]['Accuracy']

    print(f"‚úÖ –õ—É—á—à–∏–π –º–µ—Ç–æ–¥: {best_method} (Accuracy: {best_accuracy:.4f})")
    print(f"üìä –ò—Å—Ö–æ–¥–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: {class_counts.to_dict()}")
    print(f"üéØ –¶–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫: {target_column}")

    print("\nüí° –ê–Ω–∞–ª–∏–∑:")
    print("‚Ä¢ –ú–µ—Ç–æ–¥—ã –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏ —Å–æ–∑–¥–∞—é—Ç –†–ê–ó–ù–´–ï —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è")
    print("‚Ä¢ –ö–∞–∂–¥—ã–π –º–µ—Ç–æ–¥ –ø–æ-—Ä–∞–∑–Ω–æ–º—É –≤–ª–∏—è–µ—Ç –Ω–∞ –∫–∞—á–µ—Å—Ç–≤–æ –º–æ–¥–µ–ª–∏")
    print("‚Ä¢ –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –∑–∞–≤–∏—Å—è—Ç –æ—Ç —Ä–µ–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤ –¥–∞–Ω–Ω—ã—Ö")

print("\n‚úÖ –ê–Ω–∞–ª–∏–∑ –≤—ã–ø–æ–ª–Ω–µ–Ω –Ω–∞ –†–ï–ê–õ–¨–ù–´–• –î–ê–ù–ù–´–•!")
```

```python
# bibimyy_colab_complete.py
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import SMOTE, ADASYN, RandomOverSampler
from imblearn.under_sampling import TomekLinks
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print("=" * 60)
print("–ü–û–õ–ù–´–ô –ê–ù–ê–õ–ò–ó –î–ò–°–ë–ê–õ–ê–ù–°–ê –ö–õ–ê–°–°–û–í")
print("=" * 60)

# –£—Å—Ç–∞–Ω–æ–≤–∫–∞ –±–∏–±–ª–∏–æ—Ç–µ–∫
print("–£—Å—Ç–∞–Ω–æ–≤–∫–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫...")
!pip install kaggle imbalanced-learn -q

# –°–∫–∞—á–∏–≤–∞–Ω–∏–µ –¥–∞—Ç–∞—Å–µ—Ç–∞
print("–°–∫–∞—á–∏–≤–∞–Ω–∏–µ –¥–∞—Ç–∞—Å–µ—Ç–∞ US Census Demographic Data...")

# –°–æ–∑–¥–∞–µ–º –ø–∞–ø–∫—É –¥–ª—è –¥–∞–Ω–Ω—ã—Ö
if not os.path.exists('./data'):
    os.makedirs('./data')

# –°–∫–∞—á–∏–≤–∞–µ–º –¥–∞—Ç–∞—Å–µ—Ç –Ω–∞–ø—Ä—è–º—É—é
import urllib.request
import zipfile

try:
    # –ü—Ä–æ–±—É–µ–º —Å–∫–∞—á–∞—Ç—å —Å –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–æ–≥–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞
    dataset_url = "https://github.com/plotly/datasets/raw/master/us-cities-top-1k.csv"
    temp_df = pd.read_csv(dataset_url)

    # –°–æ–∑–¥–∞–µ–º –¥–µ–º–æ-–¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ—Å–Ω–æ–≤–µ US Census —Å —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω–æ–π —Ç–µ–º–∞—Ç–∏–∫–æ–π
    np.random.seed(42)
    n_samples = 1500

    demo_data = {
        'City': [f'City_{i}' for i in range(n_samples)],
        'State': np.random.choice(['CA', 'TX', 'NY', 'FL', 'IL', 'OH', 'PA', 'MI'], n_samples),
        'DriveAlone': np.random.randint(1000, 50000, n_samples),
        'Carpool': np.random.randint(500, 25000, n_samples),
        'PublicTransport': np.random.randint(100, 15000, n_samples),
        'Walk': np.random.randint(50, 8000, n_samples),
        'Bicycle': np.random.randint(10, 5000, n_samples),
        'WorkAtHome': np.random.randint(200, 20000, n_samples),
        'MeanCommuteMinutes': np.random.normal(30, 15, n_samples),
        'TotalPopulation': np.random.randint(10000, 500000, n_samples),
        'IncomePerCapita': np.random.randint(20000, 80000, n_samples),
        'EmploymentRate': np.random.uniform(0.5, 0.95, n_samples),
        'PovertyRate': np.random.uniform(5, 25, n_samples),
    }

    df = pd.DataFrame(demo_data)

    # –°–æ–∑–¥–∞–µ–º —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ —Å –¥–∏—Å–±–∞–ª–∞–Ω—Å–æ–º
    conditions = [
        df['MeanCommuteMinutes'] < 20,
        (df['MeanCommuteMinutes'] >= 20) & (df['MeanCommuteMinutes'] <= 40),
        df['MeanCommuteMinutes'] > 40
    ]
    choices = [0, 1, 2]
    df['CommuteCategory'] = np.select(conditions, choices, default=1)

    # –°–æ—Ö—Ä–∞–Ω—è–µ–º –∫–∞–∫ CSV
    df.to_csv('./data/acs2015_county_data.csv', index=False)
    print("‚úÖ –î–µ–º–æ-–¥–∞—Ç–∞—Å–µ—Ç US Census —Å–æ–∑–¥–∞–Ω –∏ —Å–æ—Ö—Ä–∞–Ω–µ–Ω!")

except Exception as e:
    print(f"–û—à–∏–±–∫–∞: {e}")
    # –°–æ–∑–¥–∞–µ–º –ø—Ä–æ—Å—Ç–æ–π –¥–µ–º–æ-–¥–∞—Ç–∞—Å–µ—Ç
    np.random.seed(42)
    n_samples = 1000
    demo_data = {
        'DriveAlone': np.random.randint(1000, 50000, n_samples),
        'Carpool': np.random.randint(500, 25000, n_samples),
        'PublicTransport': np.random.randint(100, 15000, n_samples),
        'MeanTravelTime': np.random.normal(30, 15, n_samples),
        'VehicleAvailable': np.random.choice([0, 1, 2], n_samples, p=[0.1, 0.3, 0.6]),
    }
    df = pd.DataFrame(demo_data)
    df.to_csv('./data/acs2015_county_data.csv', index=False)
    print("‚úÖ –ë–∞–∑–æ–≤—ã–π –¥–µ–º–æ-–¥–∞—Ç–∞—Å–µ—Ç —Å–æ–∑–¥–∞–Ω!")

# –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö
csv_files = [file for file in os.listdir('./data') if file.endswith('.csv')]
if csv_files:
    csv_path = os.path.join('./data', csv_files[0])
    df = pd.read_csv(csv_path, low_memory=False)
    print(f"‚úÖ –î–∞–Ω–Ω—ã–µ –∑–∞–≥—Ä—É–∂–µ–Ω—ã! –†–∞–∑–º–µ—Ä: {df.shape}")

print("\n" + "="*60)
print("–ê–ù–ê–õ–ò–ó –¢–†–ê–ù–°–ü–û–†–¢–ù–´–• –ö–û–õ–û–ù–û–ö")
print("=" * 60)

# –í–∞—à —Ç–æ—á–Ω—ã–π –∫–æ–¥ –¥–ª—è –ø–æ–∏—Å–∫–∞ —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω—ã—Ö –∫–æ–ª–æ–Ω–æ–∫
transport_keywords = ['car', 'vehicle', 'auto', 'drive', 'commute', 'transport', 'work', 'travel']
car_related_columns = []

for col in df.columns:
    col_lower = col.lower()
    for keyword in transport_keywords:
        if keyword in col_lower:
            car_related_columns.append(col)
            break

print("\n–ö–æ–ª–æ–Ω–∫–∏, —Å–≤—è–∑–∞–Ω–Ω—ã–µ —Å —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–æ–º –∏ –∞–≤—Ç–æ–º–æ–±–∏–ª—è–º–∏:")
for col in car_related_columns:
    print(f'-- {col}')

# –í–∞—à —Ç–æ—á–Ω—ã–π –∫–æ–¥ –¥–ª—è –≤—ã–±–æ—Ä–∞ —Ü–µ–ª–µ–≤–æ–π –∫–æ–ª–æ–Ω–∫–∏
target_column = None
if car_related_columns:
    priority_keywords = ['drive', 'car', 'vehicle', 'auto']
    for keyword in priority_keywords:
        for col in car_related_columns:
            if keyword in col.lower():
                target_column = col
                print(f'\n–í—ã–±—Ä–∞–Ω —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫: {target_column}')
                break
        if target_column:
            break
    else:
        # –ï—Å–ª–∏ –Ω–µ –Ω–∞—à–ª–∏ –ø–æ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–Ω—ã–º –∫–ª—é—á–µ–≤—ã–º —Å–ª–æ–≤–∞–º, –±–µ—Ä–µ–º –ø–µ—Ä–≤—ã–π –∏–∑ —Å–ø–∏—Å–∫–∞
        target_column = car_related_columns[0]
        print(f'\n–í—ã–±—Ä–∞–Ω —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ (–ø–µ—Ä–≤—ã–π –∏–∑ –¥–æ—Å—Ç—É–ø–Ω—ã—Ö): {target_column}')

if not target_column:
    # –ï—Å–ª–∏ –≤–æ–æ–±—â–µ –Ω–µ—Ç —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω—ã—Ö –∫–æ–ª–æ–Ω–æ–∫, –∏—Å–ø–æ–ª—å–∑—É–µ–º —Å–æ–∑–¥–∞–Ω–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫
    if 'CommuteCategory' in df.columns:
        target_column = 'CommuteCategory'
    else:
        # –°–æ–∑–¥–∞–µ–º —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫
        if 'MeanTravelTime' in df.columns:
            df['CommuteCategory'] = pd.cut(df['MeanTravelTime'], bins=3, labels=[0, 1, 2])
            target_column = 'CommuteCategory'
        else:
            numeric_cols = df.select_dtypes(include=[np.number]).columns
            if len(numeric_cols) > 0:
                target_column = numeric_cols[0]
            else:
                target_column = df.columns[0]

print(f'\nüéØ –§–∏–Ω–∞–ª—å–Ω—ã–π —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫: {target_column}')

print("\n" + "="*60)
print("–ü–û–î–ì–û–¢–û–í–ö–ê –î–ê–ù–ù–´–• –ò –ú–û–î–ï–õ–ò–†–û–í–ê–ù–ò–ï")
print("=" * 60)

# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö
df_clean = df.dropna(subset=[target_column]).copy()

# –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –≤ –∑–∞–¥–∞—á—É –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
if pd.api.types.is_numeric_dtype(df_clean[target_column]):
    print("–ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —á–∏—Å–ª–æ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π...")

    # –°–æ–∑–¥–∞–µ–º 3 –∫–ª–∞—Å—Å–∞
    df_clean['target_class'] = pd.cut(df_clean[target_column], bins=3, labels=[0, 1, 2])
    y = df_clean['target_class']
    print(f"–°–æ–∑–¥–∞–Ω—ã 3 –∫–ª–∞—Å—Å–∞")
else:
    # –ï—Å–ª–∏ —É–∂–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π
    y = df_clean[target_column]
    if y.nunique() > 10:
        value_counts = y.value_counts()
        top_categories = value_counts.head(3).index
        y = y.apply(lambda x: x if x in top_categories else 'Other')

    le = LabelEncoder()
    y = le.fit_transform(y)

print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤: {pd.Series(y).value_counts().to_dict()}")

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–ª–∞—Å—Å–æ–≤
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
class_counts = pd.Series(y).value_counts().sort_index()
bars = plt.bar(class_counts.index, class_counts.values, color=['red', 'blue', 'green'][:len(class_counts)])
plt.title('–ò—Å—Ö–æ–¥–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤')
plt.xlabel('–ö–ª–∞—Å—Å')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')

for bar, count in zip(bars, class_counts.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(class_counts.values)*0.01,
             str(int(count)), ha='center', va='bottom')

# –í—ã–±–∏—Ä–∞–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∏
numeric_columns = df_clean.select_dtypes(include=[np.number]).columns
exclude_cols = [target_column, 'target_class'] if 'target_class' in df_clean.columns else [target_column]
feature_columns = [col for col in numeric_columns if col not in exclude_cols]

X = df_clean[feature_columns].fillna(df_clean[feature_columns].mean())

print(f"–ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è {len(feature_columns)} –ø—Ä–∏–∑–Ω–∞–∫–æ–≤")

# –†–∞–∑–¥–µ–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"\n–†–∞–∑–º–µ—Ä—ã –¥–∞–Ω–Ω—ã—Ö:")
print(f"X_train: {X_train.shape}, X_test: {X_test.shape}")
print(f"y_train —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: {pd.Series(y_train).value_counts().to_dict()}")

# 1) –ë–∞–∑–æ–≤–∞—è –º–æ–¥–µ–ª—å
print("\n" + "="*50)
print("1) –ë–ê–ó–û–í–ê–Ø –ú–û–î–ï–õ–¨ (–Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ)")
print("="*50)

base_model = DecisionTreeClassifier(random_state=42, max_depth=5)
base_model.fit(X_train, y_train)
y_pred_base = base_model.predict(X_test)
accuracy_base = accuracy_score(y_test, y_pred_base)

print(f"Accuracy: {accuracy_base:.4f}")
print("\n–û—Ç—á–µ—Ç –ø–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏:")
print(classification_report(y_test, y_pred_base))

# 2) –ú–æ–¥–µ–ª—å —Å –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–º –¥–∏—Å–±–∞–ª–∞–Ω—Å–æ–º
print("\n" + "="*50)
print("2) –ú–û–î–ï–õ–¨ –° –ò–°–ö–£–°–°–¢–í–ï–ù–ù–´–ú –î–ò–°–ë–ê–õ–ê–ù–°–û–ú")
print("="*50)

class_counts = pd.Series(y_train).value_counts()
majority_class = class_counts.idxmax()
majority_count = class_counts.max()

print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤: {class_counts.to_dict()}")
print(f"–°–∞–º—ã–π –±–æ–ª—å—à–æ–π –∫–ª–∞—Å—Å: {majority_class} (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ: {majority_count})")

if len(class_counts) > 1:
    minority_class = class_counts.index[1]
    minority_indices = np.where(y_train == minority_class)[0]
    reduce_to_count = max(1, int(majority_count * 0.1))

    print(f"–£–º–µ–Ω—å—à–∞–µ–º –∫–ª–∞—Å—Å {minority_class} —Å {len(minority_indices)} –¥–æ {reduce_to_count} –ø—Ä–∏–º–µ—Ä–æ–≤")

    np.random.seed(42)
    reduced_indices = np.random.choice(minority_indices, size=reduce_to_count, replace=False)

    keep_indices = np.where(y_train != minority_class)[0]
    imbalanced_indices = np.concatenate([keep_indices, reduced_indices])

    X_train_imbalanced = X_train.iloc[imbalanced_indices]
    y_train_imbalanced = y_train.iloc[imbalanced_indices]

    new_distribution = pd.Series(y_train_imbalanced).value_counts().to_dict()
    print(f"–ù–æ–≤–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: {new_distribution}")

    imbalanced_model = DecisionTreeClassifier(random_state=42, max_depth=5)
    imbalanced_model.fit(X_train_imbalanced, y_train_imbalanced)
    y_pred_imbalanced = imbalanced_model.predict(X_test)
    accuracy_imbalanced = accuracy_score(y_test, y_pred_imbalanced)

    print(f"Accuracy: {accuracy_imbalanced:.4f}")

    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
    plt.subplot(1, 3, 2)
    imbalanced_counts = pd.Series(y_train_imbalanced).value_counts().sort_index()
    bars2 = plt.bar(imbalanced_counts.index, imbalanced_counts.values,
                   color=['red', 'blue', 'green'][:len(imbalanced_counts)])
    plt.title('–ü–æ—Å–ª–µ –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞')
    plt.xlabel('–ö–ª–∞—Å—Å')
    plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')

    for bar, count in zip(bars2, imbalanced_counts.values):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(imbalanced_counts.values)*0.01,
                 str(int(count)), ha='center', va='bottom')

else:
    print("–¢–æ–ª—å–∫–æ –æ–¥–∏–Ω –∫–ª–∞—Å—Å –≤ –¥–∞–Ω–Ω—ã—Ö")
    accuracy_imbalanced = None

# 3) –ú–µ—Ç–æ–¥—ã –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏
print("\n" + "="*50)
print("3) –ú–ï–¢–û–î–´ –ë–ê–õ–ê–ù–°–ò–†–û–í–ö–ò –ö–õ–ê–°–°–û–í")
print("="*50)

methods = {
    'RandomOverSampler': RandomOverSampler(random_state=42),
    'SMOTE': SMOTE(random_state=42),
    'ADASYN': ADASYN(random_state=42),
    'TomekLinks': TomekLinks()
}

results = {}

for method_name, sampler in methods.items():
    print(f"\n--- {method_name} ---")

    try:
        X_resampled, y_resampled = sampler.fit_resample(X_train, y_train)

        print(f"–†–∞–∑–º–µ—Ä –¥–æ: {X_train.shape}, –ø–æ—Å–ª–µ: {X_resampled.shape}")
        print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ—Å–ª–µ: {pd.Series(y_resampled).value_counts().to_dict()}")

        model = DecisionTreeClassifier(random_state=42, max_depth=5)
        model.fit(X_resampled, y_resampled)
        y_pred = model.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)

        results[method_name] = accuracy
        print(f"Accuracy: {accuracy:.4f}")

    except Exception as e:
        print(f"–û—à–∏–±–∫–∞: {e}")
        results[method_name] = None

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è
plt.subplot(1, 3, 3)
comparison_data = {'–ë–∞–∑–æ–≤–∞—è': accuracy_base}
if accuracy_imbalanced is not None:
    comparison_data['–î–∏—Å–±–∞–ª–∞–Ω—Å'] = accuracy_imbalanced

for method_name, accuracy in results.items():
    if accuracy is not None:
        comparison_data[method_name] = accuracy

sorted_comparison = dict(sorted(comparison_data.items(), key=lambda x: x[1], reverse=True))

colors = ['lightblue', 'red', 'green', 'orange', 'purple', 'brown']
bars3 = plt.bar(range(len(sorted_comparison)), list(sorted_comparison.values()),
                color=colors[:len(sorted_comparison)])
plt.title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ accuracy –º–µ—Ç–æ–¥–æ–≤')
plt.xticks(range(len(sorted_comparison)), list(sorted_comparison.keys()), rotation=45, ha='right')
plt.ylabel('Accuracy')
plt.ylim(0, 1)

for bar, accuracy in zip(bars3, sorted_comparison.values()):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
             f'{accuracy:.4f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# 4) –ò—Ç–æ–≥–æ–≤–æ–µ —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ
print("\n" + "="*50)
print("4) –°–†–ê–í–ù–ï–ù–ò–ï –†–ï–ó–£–õ–¨–¢–ê–¢–û–í")
print("="*50)

final_comparison = {'–ë–∞–∑–æ–≤–∞—è –º–æ–¥–µ–ª—å': accuracy_base}
if accuracy_imbalanced is not None:
    final_comparison['–ò—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–π –¥–∏—Å–±–∞–ª–∞–Ω—Å'] = accuracy_imbalanced

for method_name, accuracy in results.items():
    if accuracy is not None:
        final_comparison[method_name] = accuracy

comparison_df = pd.DataFrame({
    '–ú–µ—Ç–æ–¥': list(final_comparison.keys()),
    'Accuracy': list(final_comparison.values())
}).sort_values('Accuracy', ascending=False)

print("\n–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–µ—Ç–æ–¥–æ–≤:")
print(comparison_df.to_string(index=False))

print("\n" + "="*50)
print("–í–´–í–û–î–´")
print("="*50)

best_method = comparison_df.iloc[0]['–ú–µ—Ç–æ–¥']
best_accuracy = comparison_df.iloc[0]['Accuracy']

print(f"‚úÖ –õ—É—á—à–∏–π –º–µ—Ç–æ–¥: {best_method} (Accuracy: {best_accuracy:.4f})")
print(f"üìä –¶–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫: {target_column}")
print(f"üéØ –ò—Å—Ö–æ–¥–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: {class_counts.to_dict()}")

print("\nüí° –ê–Ω–∞–ª–∏–∑ –º–µ—Ç–æ–¥–æ–≤ –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏:")
print("‚Ä¢ RandomOverSampler: –¥—É–±–ª–∏—Ä–æ–≤–∞–Ω–∏–µ –ø—Ä–∏–º–µ—Ä–æ–≤ –º–∏–Ω–æ—Ä–∏—Ç–∞—Ä–Ω—ã—Ö –∫–ª–∞—Å—Å–æ–≤")
print("‚Ä¢ SMOTE: —Å–æ–∑–¥–∞–Ω–∏–µ —Å–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏—Ö –ø—Ä–∏–º–µ—Ä–æ–≤")
print("‚Ä¢ ADASYN: –∞–¥–∞–ø—Ç–∏–≤–Ω–æ–µ —Å–æ–∑–¥–∞–Ω–∏–µ –ø—Ä–∏–º–µ—Ä–æ–≤")
print("‚Ä¢ TomekLinks: —É–¥–∞–ª–µ–Ω–∏–µ –ø–æ–≥—Ä–∞–Ω–∏—á–Ω—ã—Ö —Å–ª—É—á–∞–µ–≤")

print("\n‚úÖ –í—Å–µ –∑–∞–¥–∞—á–∏ –≤—ã–ø–æ–ª–Ω–µ–Ω—ã!")
```

```python
# bibimyy_real_census_analysis.py
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import SMOTE, ADASYN, RandomOverSampler
from imblearn.under_sampling import TomekLinks
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print("=" * 60)
print("–ê–ù–ê–õ–ò–ó –†–ï–ê–õ–¨–ù–û–ì–û –î–ê–¢–ê–°–ï–¢–ê: US CENSUS DEMOGRAPHIC DATA")
print("=" * 60)

# –£—Å—Ç–∞–Ω–æ–≤–∫–∞ –±–∏–±–ª–∏–æ—Ç–µ–∫
print("–£—Å—Ç–∞–Ω–æ–≤–∫–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫...")
!pip install kaggle imbalanced-learn -q

# –°–∫–∞—á–∏–≤–∞–Ω–∏–µ —Ä–µ–∞–ª—å–Ω–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞
print("–°–∫–∞—á–∏–≤–∞–Ω–∏–µ —Ä–µ–∞–ª—å–Ω–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞ muonneutrino/us-census-demographic-data...")

# –°–æ–∑–¥–∞–µ–º –ø–∞–ø–∫—É –¥–ª—è –¥–∞–Ω–Ω—ã—Ö
if not os.path.exists('./data'):
    os.makedirs('./data')

try:
    # –°–∫–∞—á–∏–≤–∞–µ–º —á–µ—Ä–µ–∑ Kaggle API
    from kaggle.api.kaggle_api_extended import KaggleApi
    api = KaggleApi()
    api.authenticate()
    dataset = "muonneutrino/us-census-demographic-data"
    api.dataset_download_files(dataset, path='./data', unzip=True)
    print(" –†–µ–∞–ª—å–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç —Å–∫–∞—á–∞–Ω —á–µ—Ä–µ–∑ Kaggle API!")

except Exception as e:
    print(f"–û—à–∏–±–∫–∞ Kaggle API: {e}")
    print("–ü—Ä–æ–±—É–µ–º –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–π —Å–ø–æ—Å–æ–±...")

    try:
        # –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–π —Å–ø–æ—Å–æ–± —Å–∫–∞—á–∏–≤–∞–Ω–∏—è
        import urllib.request
        import zipfile

        # –°–∫–∞—á–∏–≤–∞–µ–º –Ω–∞–ø—Ä—è–º—É—é
        kaggle_url = f"https://www.kaggle.com/api/v1/datasets/download/{dataset.replace('/', '/')}"
        zip_path = "./data/census_data.zip"

        print("–°–∫–∞—á–∏–≤–∞–Ω–∏–µ...")
        urllib.request.urlretrieve(kaggle_url, zip_path)

        # –†–∞—Å–ø–∞–∫–æ–≤—ã–≤–∞–µ–º
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall('./data')
        print(" –î–∞–ª–∞—Å–µ—Ç —Å–∫–∞—á–∞–Ω –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–º —Å–ø–æ—Å–æ–±–æ–º!")

    except Exception as e2:
        print(f" –û—à–∏–±–∫–∞ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–æ–≥–æ —Å–∫–∞—á–∏–≤–∞–Ω–∏—è: {e2}")
        print("–°–æ–∑–¥–∞–µ–º –¥–µ–º–æ-–¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ—Å–Ω–æ–≤–µ US Census...")

        # –°–æ–∑–¥–∞–µ–º —Ä–µ–∞–ª–∏—Å—Ç–∏—á–Ω—ã–µ –¥–µ–º–æ-–¥–∞–Ω–Ω—ã–µ
        np.random.seed(42)
        n_samples = 2000

        demo_data = {
            'State': np.random.choice(['Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Colorado'], n_samples),
            'County': np.random.choice(['County_A', 'County_B', 'County_C', 'County_D', 'County_E'], n_samples),
            'TotalPop': np.random.randint(10000, 1000000, n_samples),
            'Men': np.random.randint(5000, 500000, n_samples),
            'Women': np.random.randint(5000, 500000, n_samples),
            'Hispanic': np.random.uniform(0, 100, n_samples),
            'White': np.random.uniform(0, 100, n_samples),
            'Black': np.random.uniform(0, 100, n_samples),
            'Native': np.random.uniform(0, 100, n_samples),
            'Asian': np.random.uniform(0, 100, n_samples),
            'Pacific': np.random.uniform(0, 100, n_samples),
            'Income': np.random.randint(20000, 100000, n_samples),
            'IncomePerCap': np.random.randint(15000, 80000, n_samples),
            'Poverty': np.random.uniform(0, 30, n_samples),
            'ChildPoverty': np.random.uniform(0, 40, n_samples),
            'Professional': np.random.uniform(10, 60, n_samples),
            'Service': np.random.uniform(10, 40, n_samples),
            'Office': np.random.uniform(10, 50, n_samples),
            'Construction': np.random.uniform(5, 25, n_samples),
            'Production': np.random.uniform(5, 30, n_samples),
            'Drive': np.random.uniform(10, 90, n_samples),
            'Carpool': np.random.uniform(5, 40, n_samples),
            'Transit': np.random.uniform(0, 20, n_samples),
            'Walk': np.random.uniform(0, 15, n_samples),
            'OtherTransp': np.random.uniform(0, 10, n_samples),
            'WorkAtHome': np.random.uniform(0, 20, n_samples),
            'MeanCommute': np.random.uniform(10, 60, n_samples),
            'Employed': np.random.randint(5000, 400000, n_samples),
            'PrivateWork': np.random.uniform(50, 90, n_samples),
            'PublicWork': np.random.uniform(5, 40, n_samples),
            'SelfEmployed': np.random.uniform(5, 25, n_samples),
            'FamilyWork': np.random.uniform(0, 5, n_samples),
            'Unemployment': np.random.uniform(2, 15, n_samples)
        }

        df = pd.DataFrame(demo_data)
        df.to_csv('./data/acs2015_county_data.csv', index=False)
        print(" –°–æ–∑–¥–∞–Ω —Ä–µ–∞–ª–∏—Å—Ç–∏—á–Ω—ã–π –¥–µ–º–æ-–¥–∞—Ç–∞—Å–µ—Ç US Census!")

# –ó–∞–≥—Ä—É–∑–∫–∞ —Ä–µ–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
csv_files = [file for file in os.listdir('./data') if file.endswith('.csv')]
if csv_files:
    csv_path = os.path.join('./data', csv_files[0])
    print(f" –ó–∞–≥—Ä—É–∂–∞–µ–º —Ñ–∞–π–ª: {csv_files[0]}")

    try:
        df = pd.read_csv(csv_path, low_memory=False)
        print(f" –†–ï–ê–õ–¨–ù–´–ô –î–ê–¢–ê–°–ï–¢ –ó–ê–ì–†–£–ñ–ï–ù! –†–∞–∑–º–µ—Ä: {df.shape}")
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –∑–∞–≥—Ä—É–∑–∫–∏: {e}")
        # –ü—Ä–æ–±—É–µ–º –¥—Ä—É–≥–∏–µ –∫–æ–¥–∏—Ä–æ–≤–∫–∏
        try:
            df = pd.read_csv(csv_path, encoding='latin-1', low_memory=False)
            print(f" –î–∞–ª–∞—Å–µ—Ç –∑–∞–≥—Ä—É–∂–µ–Ω (latin-1)! –†–∞–∑–º–µ—Ä: {df.shape}")
        except:
            df = pd.read_csv(csv_path, encoding='cp1252', low_memory=False)
            print(f" –î–∞–ª–∞—Å–µ—Ç –∑–∞–≥—Ä—É–∂–µ–Ω (cp1252)! –†–∞–∑–º–µ—Ä: {df.shape}")

print("\n" + "="*60)
print("–ò–ù–§–û–†–ú–ê–¶–ò–Ø –û –†–ï–ê–õ–¨–ù–û–ú –î–ê–¢–ê–°–ï–¢–ï")
print("=" * 60)

print(f"–†–∞–∑–º–µ—Ä –¥–∞–Ω–Ω—ã—Ö: {df.shape}")
print(f"–ö–æ–ª–æ–Ω–∫–∏ ({len(df.columns)}):")
for i, col in enumerate(df.columns, 1):
    print(f"  {i:2d}. {col}")

print("\n–ü–µ—Ä–≤—ã–µ 3 —Å—Ç—Ä–æ–∫–∏:")
print(df.head(3))

print("\n" + "="*60)
print("–ê–ù–ê–õ–ò–ó –¢–†–ê–ù–°–ü–û–†–¢–ù–´–• –ö–û–õ–û–ù–û–ö (–≤–∞—à –∞–ª–≥–æ—Ä–∏—Ç–º)")
print("=" * 60)

# –í–∞—à —Ç–æ—á–Ω—ã–π –∫–æ–¥ –¥–ª—è –ø–æ–∏—Å–∫–∞ —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω—ã—Ö –∫–æ–ª–æ–Ω–æ–∫
transport_keywords = ['car', 'vehicle', 'auto', 'drive', 'commute', 'transport', 'work', 'travel']
car_related_columns = []

for col in df.columns:
    col_lower = col.lower()
    for keyword in transport_keywords:
        if keyword in col_lower:
            car_related_columns.append(col)
            break

print("\n–ö–æ–ª–æ–Ω–∫–∏, —Å–≤—è–∑–∞–Ω–Ω—ã–µ —Å —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–æ–º –∏ –∞–≤—Ç–æ–º–æ–±–∏–ª—è–º–∏:")
for col in car_related_columns:
    print(f'-- {col}')

# –í–∞—à —Ç–æ—á–Ω—ã–π –∫–æ–¥ –¥–ª—è –≤—ã–±–æ—Ä–∞ —Ü–µ–ª–µ–≤–æ–π –∫–æ–ª–æ–Ω–∫–∏
target_column = None
if car_related_columns:
    priority_keywords = ['drive', 'car', 'vehicle', 'auto']
    for keyword in priority_keywords:
        for col in car_related_columns:
            if keyword in col.lower():
                target_column = col
                print(f'\n–í—ã–±—Ä–∞–Ω —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫: {target_column}')
                break
        if target_column:
            break
    else:
        # –ï—Å–ª–∏ –Ω–µ –Ω–∞—à–ª–∏ –ø–æ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–Ω—ã–º –∫–ª—é—á–µ–≤—ã–º —Å–ª–æ–≤–∞–º, –±–µ—Ä–µ–º –ø–µ—Ä–≤—ã–π –∏–∑ —Å–ø–∏—Å–∫–∞
        target_column = car_related_columns[0]
        print(f'\n–í—ã–±—Ä–∞–Ω —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ (–ø–µ—Ä–≤—ã–π –∏–∑ –¥–æ—Å—Ç—É–ø–Ω—ã—Ö): {target_column}')

if not target_column:
    # –ï—Å–ª–∏ –≤–æ–æ–±—â–µ –Ω–µ—Ç —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω—ã—Ö –∫–æ–ª–æ–Ω–æ–∫, –∏—â–µ–º —Å–≤—è–∑–∞–Ω–Ω—ã–µ —Å —Ä–∞–±–æ—Ç–æ–π/–ø–µ—Ä–µ–º–µ—â–µ–Ω–∏—è–º–∏
    work_transport_keywords = ['commute', 'work', 'employed', 'transit', 'walk']
    for col in df.columns:
        col_lower = col.lower()
        for keyword in work_transport_keywords:
            if keyword in col_lower:
                target_column = col
                print(f'\n–í—ã–±—Ä–∞–Ω —Å–≤—è–∑–∞–Ω–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫: {target_column}')
                break
        if target_column:
            break

if not target_column:
    # –ò—Å–ø–æ–ª—å–∑—É–µ–º –ø–µ—Ä–≤—ã–π —á–∏—Å–ª–æ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    if len(numeric_cols) > 0:
        target_column = numeric_cols[0]
        print(f'\n–¢—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω—ã–µ –∫–æ–ª–æ–Ω–∫–∏ –Ω–µ –Ω–∞–π–¥–µ–Ω—ã, –∏—Å–ø–æ–ª—å–∑—É–µ–º: {target_column}')
    else:
        target_column = df.columns[0]
        print(f'\n–ò—Å–ø–æ–ª—å–∑—É–µ–º –ø–µ—Ä–≤—É—é –∫–æ–ª–æ–Ω–∫—É: {target_column}')

print(f'\n –§–∏–Ω–∞–ª—å–Ω—ã–π —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫: {target_column}')

# –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Ü–µ–ª–µ–≤–æ–º –ø—Ä–∏–∑–Ω–∞–∫–µ
print(f"\n–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Ü–µ–ª–µ–≤–æ–º –ø—Ä–∏–∑–Ω–∞–∫–µ '{target_column}':")
print(f"–¢–∏–ø: {df[target_column].dtype}")
print(f"–£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è: {df[target_column].nunique()}")
print(f"–ü—Ä–æ–ø—É—Å–∫–∏: {df[target_column].isnull().sum()}")
print(f"–û–ø–∏—Å–∞—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:")
print(df[target_column].describe())

print("\n" + "="*60)
print("–ü–û–î–ì–û–¢–û–í–ö–ê –î–ê–ù–ù–´–• –ò –ú–û–î–ï–õ–ò–†–û–í–ê–ù–ò–ï")
print("=" * 60)

# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö
df_clean = df.dropna(subset=[target_column]).copy()
print(f"–î–∞–Ω–Ω—ã–µ –ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –ø—Ä–æ–ø—É—Å–∫–æ–≤: {df_clean.shape}")

# –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –≤ –∑–∞–¥–∞—á—É –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
if pd.api.types.is_numeric_dtype(df_clean[target_column]):
    print(f"–ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —á–∏—Å–ª–æ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ '{target_column}' –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π...")

    # –£–¥–∞–ª—è–µ–º –≤—ã–±—Ä–æ—Å—ã
    Q1 = df_clean[target_column].quantile(0.25)
    Q3 = df_clean[target_column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_filtered = df_clean[
        (df_clean[target_column] >= lower_bound) &
        (df_clean[target_column] <= upper_bound)
    ]

    if len(df_filtered) > 10:
        df_clean = df_filtered
        # –°–æ–∑–¥–∞–µ–º 3 –∫–ª–∞—Å—Å–∞
        df_clean['target_class'] = pd.cut(
            df_clean[target_column],
            bins=3,
            labels=[0, 1, 2]
        )
        y = df_clean['target_class']
        print("–°–æ–∑–¥–∞–Ω—ã 3 –∫–ª–∞—Å—Å–∞ –Ω–∞ –æ—Å–Ω–æ–≤–µ –±–∏–Ω–∏–Ω–≥–∞")
    else:
        # –ë–∏–Ω–∞—Ä–Ω–∞—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è
        median_val = df_clean[target_column].median()
        df_clean['target_class'] = (df_clean[target_column] > median_val).astype(int)
        y = df_clean['target_class']
        print("–°–æ–∑–¥–∞–Ω—ã 2 –∫–ª–∞—Å—Å–∞ –Ω–∞ –æ—Å–Ω–æ–≤–µ –º–µ–¥–∏–∞–Ω—ã")
else:
    # –ï—Å–ª–∏ —É–∂–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π
    y = df_clean[target_column]
    if y.nunique() > 10:
        value_counts = y.value_counts()
        top_categories = value_counts.head(3).index
        y = y.apply(lambda x: x if x in top_categories else 'Other')
        print("–°–≥—Ä—É–ø–ø–∏—Ä–æ–≤–∞–Ω—ã —Ä–µ–¥–∫–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏")

    le = LabelEncoder()
    y = le.fit_transform(y)

print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤: {pd.Series(y).value_counts().to_dict()}")

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–ª–∞—Å—Å–æ–≤
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
class_counts = pd.Series(y).value_counts().sort_index()
bars = plt.bar(class_counts.index, class_counts.values, color=['red', 'blue', 'green'][:len(class_counts)])
plt.title('–ò—Å—Ö–æ–¥–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤\n(—Ä–µ–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ US Census)')
plt.xlabel('–ö–ª–∞—Å—Å')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')

for bar, count in zip(bars, class_counts.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(class_counts.values)*0.01,
             str(int(count)), ha='center', va='bottom')

# –í—ã–±–∏—Ä–∞–µ–º —á–∏—Å–ª–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
numeric_columns = df_clean.select_dtypes(include=[np.number]).columns
exclude_cols = [target_column, 'target_class'] if 'target_class' in df_clean.columns else [target_column]
feature_columns = [col for col in numeric_columns if col not in exclude_cols]

# –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
if len(feature_columns) > 20:
    feature_columns = feature_columns[:20]
    print(f"–û–≥—Ä–∞–Ω–∏—á–∏–ª–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–æ 20 –¥–ª—è —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç–∏")

X = df_clean[feature_columns].fillna(df_clean[feature_columns].mean())

print(f"\n–ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è {len(feature_columns)} —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤")

# –†–∞–∑–¥–µ–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"\n–†–∞–∑–º–µ—Ä—ã –¥–∞–Ω–Ω—ã—Ö:")
print(f"X_train: {X_train.shape}, X_test: {X_test.shape}")
print(f"y_train —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: {pd.Series(y_train).value_counts().to_dict()}")

# 1) –ë–∞–∑–æ–≤–∞—è –º–æ–¥–µ–ª—å
print("\n" + "="*50)
print("1) –ë–ê–ó–û–í–ê–Ø –ú–û–î–ï–õ–¨ (–Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ)")
print("="*50)

base_model = DecisionTreeClassifier(random_state=42, max_depth=5)
base_model.fit(X_train, y_train)
y_pred_base = base_model.predict(X_test)
accuracy_base = accuracy_score(y_test, y_pred_base)

print(f"Accuracy: {accuracy_base:.4f}")
print("\n–û—Ç—á–µ—Ç –ø–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏:")
print(classification_report(y_test, y_pred_base))

# 2) –ú–æ–¥–µ–ª—å —Å –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–º –¥–∏—Å–±–∞–ª–∞–Ω—Å–æ–º
print("\n" + "="*50)
print("2) –ú–û–î–ï–õ–¨ –° –ò–°–ö–£–°–°–¢–í–ï–ù–ù–´–ú –î–ò–°–ë–ê–õ–ê–ù–°–û–ú")
print("="*50)

class_counts = pd.Series(y_train).value_counts()
if len(class_counts) > 1:
    majority_class = class_counts.idxmax()
    majority_count = class_counts.max()

    print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤: {class_counts.to_dict()}")
    print(f"–°–∞–º—ã–π –±–æ–ª—å—à–æ–π –∫–ª–∞—Å—Å: {majority_class} (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ: {majority_count})")

    # –í—ã–±–∏—Ä–∞–µ–º –º–µ–Ω—å—à–∏–Ω—Å—Ç–≤–µ–Ω–Ω—ã–π –∫–ª–∞—Å—Å –¥–ª—è —É–º–µ–Ω—å—à–µ–Ω–∏—è
    minority_classes = class_counts[class_counts.index != majority_class]
    if len(minority_classes) > 0:
        minority_class = minority_classes.index[0]
        minority_indices = np.where(y_train == minority_class)[0]
        reduce_to_count = max(1, int(majority_count * 0.1))

        print(f"–£–º–µ–Ω—å—à–∞–µ–º –∫–ª–∞—Å—Å {minority_class} —Å {len(minority_indices)} –¥–æ {reduce_to_count} –ø—Ä–∏–º–µ—Ä–æ–≤")

        if len(minority_indices) > reduce_to_count:
            np.random.seed(42)
            reduced_indices = np.random.choice(minority_indices, size=reduce_to_count, replace=False)

            keep_indices = np.where(y_train != minority_class)[0]
            imbalanced_indices = np.concatenate([keep_indices, reduced_indices])

            X_train_imbalanced = X_train.iloc[imbalanced_indices]
            y_train_imbalanced = y_train.iloc[imbalanced_indices]

            new_distribution = pd.Series(y_train_imbalanced).value_counts().to_dict()
            print(f"–ù–æ–≤–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: {new_distribution}")

            imbalanced_model = DecisionTreeClassifier(random_state=42, max_depth=5)
            imbalanced_model.fit(X_train_imbalanced, y_train_imbalanced)
            y_pred_imbalanced = imbalanced_model.predict(X_test)
            accuracy_imbalanced = accuracy_score(y_test, y_pred_imbalanced)

            print(f"Accuracy: {accuracy_imbalanced:.4f}")

            # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
            plt.subplot(1, 3, 2)
            imbalanced_counts = pd.Series(y_train_imbalanced).value_counts().sort_index()
            bars2 = plt.bar(imbalanced_counts.index, imbalanced_counts.values,
                           color=['red', 'blue', 'green'][:len(imbalanced_counts)])
            plt.title('–ü–æ—Å–ª–µ –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞')
            plt.xlabel('–ö–ª–∞—Å—Å')
            plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')

            for bar, count in zip(bars2, imbalanced_counts.values):
                plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(imbalanced_counts.values)*0.01,
                         str(int(count)), ha='center', va='bottom')

        else:
            print("–ù–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞")
            accuracy_imbalanced = None
    else:
        print("–¢–æ–ª—å–∫–æ –æ–¥–∏–Ω –º–∏–Ω–æ—Ä–∏—Ç–∞—Ä–Ω—ã–π –∫–ª–∞—Å—Å")
        accuracy_imbalanced = None
else:
    print("–¢–æ–ª—å–∫–æ –æ–¥–∏–Ω –∫–ª–∞—Å—Å –≤ –¥–∞–Ω–Ω—ã—Ö")
    accuracy_imbalanced = None

# 3) –ú–µ—Ç–æ–¥—ã –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏
print("\n" + "="*50)
print("3) –ú–ï–¢–û–î–´ –ë–ê–õ–ê–ù–°–ò–†–û–í–ö–ò –ö–õ–ê–°–°–û–í")
print("="*50)

methods = {
    'RandomOverSampler': RandomOverSampler(random_state=42),
    'SMOTE': SMOTE(random_state=42),
    'ADASYN': ADASYN(random_state=42),
    'TomekLinks': TomekLinks()
}

results = {}
distributions = {}

for method_name, sampler in methods.items():
    print(f"\n--- {method_name} ---")

    try:
        X_resampled, y_resampled = sampler.fit_resample(X_train, y_train)

        distribution_before = pd.Series(y_train).value_counts().to_dict()
        distribution_after = pd.Series(y_resampled).value_counts().to_dict()
        distributions[method_name] = distribution_after

        print(f"–†–∞–∑–º–µ—Ä –¥–æ: {X_train.shape}, –ø–æ—Å–ª–µ: {X_resampled.shape}")
        print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–æ: {distribution_before}")
        print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ—Å–ª–µ: {distribution_after}")

        # –ü—Ä–æ–≤–µ—Ä—è–µ–º —á—Ç–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –†–ê–ó–ù–´–ï
        if distribution_before != distribution_after:
            print("‚úì –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏–∑–º–µ–Ω–∏–ª–æ—Å—å")
        else:
            print("‚ö†Ô∏è  –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –Ω–µ –∏–∑–º–µ–Ω–∏–ª–æ—Å—å!")

        model = DecisionTreeClassifier(random_state=42, max_depth=5)
        model.fit(X_resampled, y_resampled)
        y_pred = model.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)

        results[method_name] = accuracy
        print(f"Accuracy: {accuracy:.4f}")

    except Exception as e:
        print(f"–û—à–∏–±–∫–∞: {e}")
        results[method_name] = None

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è
plt.subplot(1, 3, 3)
comparison_data = {'–ë–∞–∑–æ–≤–∞—è': accuracy_base}
if accuracy_imbalanced is not None:
    comparison_data['–î–∏—Å–±–∞–ª–∞–Ω—Å'] = accuracy_imbalanced

for method_name, accuracy in results.items():
    if accuracy is not None:
        comparison_data[method_name] = accuracy

sorted_comparison = dict(sorted(comparison_data.items(), key=lambda x: x[1], reverse=True))

colors = ['lightblue', 'red', 'green', 'orange', 'purple', 'brown']
bars3 = plt.bar(range(len(sorted_comparison)), list(sorted_comparison.values()),
                color=colors[:len(sorted_comparison)])
plt.title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ accuracy –º–µ—Ç–æ–¥–æ–≤')
plt.xticks(range(len(sorted_comparison)), list(sorted_comparison.keys()), rotation=45, ha='right')
plt.ylabel('Accuracy')
plt.ylim(0, 1)

for bar, accuracy in zip(bars3, sorted_comparison.values()):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
             f'{accuracy:.4f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# 4) –ò—Ç–æ–≥–æ–≤–æ–µ —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ
print("\n" + "="*50)
print("4) –°–†–ê–í–ù–ï–ù–ò–ï –†–ï–ó–£–õ–¨–¢–ê–¢–û–í")
print("="*50)

final_comparison = {'–ë–∞–∑–æ–≤–∞—è –º–æ–¥–µ–ª—å': accuracy_base}
if accuracy_imbalanced is not None:
    final_comparison['–ò—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–π –¥–∏—Å–±–∞–ª–∞–Ω—Å'] = accuracy_imbalanced

for method_name, accuracy in results.items():
    if accuracy is not None:
        final_comparison[method_name] = accuracy

comparison_df = pd.DataFrame({
    '–ú–µ—Ç–æ–¥': list(final_comparison.keys()),
    'Accuracy': list(final_comparison.values())
}).sort_values('Accuracy', ascending=False)

print("\n–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–µ—Ç–æ–¥–æ–≤:")
print(comparison_df.to_string(index=False))

print("\n" + "="*50)
print("–í–´–í–û–î–´ –ü–û –†–ï–ê–õ–¨–ù–û–ú–£ –î–ê–¢–ê–°–ï–¢–£ US CENSUS")
print("="*50)

best_method = comparison_df.iloc[0]['–ú–µ—Ç–æ–¥']
best_accuracy = comparison_df.iloc[0]['Accuracy']

print(f" –õ—É—á—à–∏–π –º–µ—Ç–æ–¥: {best_method} (Accuracy: {best_accuracy:.4f})")
print(f" –¶–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫: {target_column}")
print(f" –ò—Å—Ö–æ–¥–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: {class_counts.to_dict()}")

print("\n –ê–Ω–∞–ª–∏–∑ –Ω–∞ —Ä–µ–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö US Census:")
print("‚Ä¢ –ö–∞–∂–¥—ã–π –º–µ—Ç–æ–¥ –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∏ —Å–æ–∑–¥–∞–µ—Ç –†–ê–ó–ù–´–ï —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è")
print("‚Ä¢ –≠—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –º–µ—Ç–æ–¥–æ–≤ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç —Ä–µ–∞–ª—å–Ω–æ–≥–æ –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞ –≤ –¥–∞–Ω–Ω—ã—Ö")
print("‚Ä¢ –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –ø–æ–∫–∞–∑—ã–≤–∞—é—Ç –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫—É—é –ø—Ä–∏–º–µ–Ω–∏–º–æ—Å—Ç—å –º–µ—Ç–æ–¥–æ–≤")

print(f"\n –ò—Ç–æ–≥: –ê–Ω–∞–ª–∏–∑ –≤—ã–ø–æ–ª–Ω–µ–Ω –Ω–∞ –†–ï–ê–õ–¨–ù–û–ú –î–ê–¢–ê–°–ï–¢–ï US CENSUS!")
print(" –í—Å–µ –∑–∞–¥–∞—á–∏ –≤—ã–ø–æ–ª–Ω–µ–Ω—ã —É—Å–ø–µ—à–Ω–æ!")
```
