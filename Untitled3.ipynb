```python
# –£—Å—Ç–∞–Ω–æ–≤–∫–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫ (–≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –æ–¥–∏–Ω —Ä–∞–∑)
!pip install requests beautifulsoup4 pandas lxml

import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime, timedelta
import time

# –ù–∞—Å—Ç—Ä–æ–π–∫–∏
BASE_URL = 'https://ria.ru'
HEADERS = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0 Safari/537.36'
}

def parse_ria_date(date_str):
    """–ü—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç –¥–∞—Ç—É –∏–∑ —Ñ–æ—Ä–º–∞—Ç–∞ –†–ò–ê –≤ datetime."""
    now = datetime.now()
    date_str = date_str.replace('\xa0', ' ').strip()

    if '–°–µ–≥–æ–¥–Ω—è' in date_str:
        time_part = date_str.split(',')[-1].strip()
        hour, minute = map(int, time_part.split(':'))
        return now.replace(hour=hour, minute=minute, second=0, microsecond=0)
    elif '–í—á–µ—Ä–∞' in date_str:
        time_part = date_str.split(',')[-1].strip()
        hour, minute = map(int, time_part.split(':'))
        yesterday = now - timedelta(days=1)
        return yesterday.replace(hour=hour, minute=minute, second=0, microsecond=0)
    else:
        months = {
            '—è–Ω–≤–∞—Ä—è': '01', '—Ñ–µ–≤—Ä–∞–ª—è': '02', '–º–∞—Ä—Ç–∞': '03', '–∞–ø—Ä–µ–ª—è': '04',
            '–º–∞—è': '05', '–∏—é–Ω—è': '06', '–∏—é–ª—è': '07', '–∞–≤–≥—É—Å—Ç–∞': '08',
            '—Å–µ–Ω—Ç—è–±—Ä—è': '09', '–æ–∫—Ç—è–±—Ä—è': '10', '–Ω–æ—è–±—Ä—è': '11', '–¥–µ–∫–∞–±—Ä—è': '12'
        }
        for ru, num in months.items():
            date_str = date_str.replace(ru, num)

        try:
            parts = date_str.split(',')
            date_part = parts[0].strip()
            time_part = parts[1].strip() if len(parts) > 1 else "00:00"

            if len(date_part.split()) == 2:  # –¥–µ–Ω—å –∏ –º–µ—Å—è—Ü
                date_part += f" {now.year}"
            full_str = f"{date_part} {time_part}"
            return datetime.strptime(full_str, "%d %m %Y %H:%M")
        except Exception as e:
            print(f"–ù–µ —É–¥–∞–ª–æ—Å—å —Ä–∞—Å–ø–∞—Ä—Å–∏—Ç—å –¥–∞—Ç—É: {date_str} ‚Üí {e}")
            return None


def get_news_list(max_news=None, since_date=None):
    """–ü–æ–ª—É—á–∞–µ—Ç —Å–ø–∏—Å–æ–∫ –Ω–æ–≤–æ—Å—Ç–µ–π —Å –≥–ª–∞–≤–Ω–æ–π —Å—Ç—Ä–∞–Ω–∏—Ü—ã –†–ò–ê."""
    response = requests.get(BASE_URL, headers=HEADERS)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'lxml')

    # –û—Å–Ω–æ–≤–Ω–æ–π —Å–µ–ª–µ–∫—Ç–æ—Ä –Ω–æ–≤–æ—Å—Ç–µ–π
    items = soup.select('a[class*="list-item"]')
    if not items:
        items = soup.select('div[cell-list__item] a')

    news_list = []
    for item in items:
        if max_news and len(news_list) >= max_news:
            break

        title_tag = item.find('div', class_='list-item__title') or item.find('span', class_='cell-list__item-title')
        if not title_tag:
            continue
        title = title_tag.get_text(strip=True)
        link = item['href']
        if not link.startswith('http'):
            link = BASE_URL + link

        date_tag = item.find('div', class_='list-item__date') or item.find('span', class_='cell-list__item-date')
        date_str = date_tag.get_text(strip=True) if date_tag else ''
        pub_date = parse_ria_date(date_str)

        if since_date and pub_date and pub_date < since_date:
            if max_news is None:
                break

        news_list.append({'title': title, 'url': link, 'date': pub_date, 'date_str': date_str})

    return news_list

def get_article_text(url):
    """–ò–∑–≤–ª–µ–∫–∞–µ—Ç —Ç–µ–∫—Å—Ç —Å—Ç–∞—Ç—å–∏ –ø–æ URL."""
    try:
        resp = requests.get(url, headers=HEADERS, timeout=10)
        soup = BeautifulSoup(resp.text, 'lxml')
        text_blocks = soup.select('div.article__text') or soup.select('div[data-type="text"]')
        text = '\n'.join([block.get_text(strip=True) for block in text_blocks])
        return text if text else "–¢–µ–∫—Å—Ç –Ω–µ –Ω–∞–π–¥–µ–Ω"
    except Exception as e:
        return f"–û—à–∏–±–∫–∞ –∑–∞–≥—Ä—É–∑–∫–∏: {str(e)}"

# =============================
#  –ü–ê–†–ê–ú–ï–¢–†–´ (–ò–ó–ú–ï–ù–ò–¢–ï –ó–î–ï–°–¨)
# =============================
MAX_NEWS = 5          # –ö–æ–ª-–≤–æ –Ω–æ–≤–æ—Å—Ç–µ–π (–æ—Å—Ç–∞–≤—å—Ç–µ None, –µ—Å–ª–∏ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç–µ –¥–∞—Ç—É)
SINCE_DATE_STR = None # –ù–∞–ø—Ä–∏–º–µ—Ä: "2025-11-01" (–æ—Å—Ç–∞–≤—å—Ç–µ None, –µ—Å–ª–∏ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç–µ MAX_NEWS)

# –ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ä–µ–∂–∏–º–∞
since_date = None
if SINCE_DATE_STR:
    since_date = datetime.strptime(SINCE_DATE_STR, "%Y-%m-%d")
    print(f"–ü–∞—Ä—Å–∏–Ω–≥ –Ω–æ–≤–æ—Å—Ç–µ–π —Å {SINCE_DATE_STR}")
else:
    print(f"–ü–∞—Ä—Å–∏–Ω–≥ {MAX_NEWS} –ø–æ—Å–ª–µ–¥–Ω–∏—Ö –Ω–æ–≤–æ—Å—Ç–µ–π")

# =============================
#  –ó–ê–ü–£–°–ö –ü–ê–†–°–ò–ù–ì–ê
# =============================
print("–ü–æ–ª—É—á–µ–Ω–∏–µ —Å–ø–∏—Å–∫–∞ –Ω–æ–≤–æ—Å—Ç–µ–π...")
news_list = get_news_list(max_news=MAX_NEWS, since_date=since_date)
print(f"–ù–∞–π–¥–µ–Ω–æ {len(news_list)} –Ω–æ–≤–æ—Å—Ç–µ–π.")

results = []
for i, news in enumerate(news_list, 1):
    print(f"[{i}/{len(news_list)}] –ó–∞–≥—Ä—É–∑–∫–∞: {news['title'][:60]}...")
    text = get_article_text(news['url'])
    results.append({
        'title': news['title'],
        'url': news['url'],
        'date': news['date'],
        'text': text
    })
    time.sleep(0.5)  # —É–≤–∞–∂–µ–Ω–∏–µ –∫ —Å–µ—Ä–≤–µ—Ä—É

# –°–æ–∑–¥–∞–Ω–∏–µ DataFrame
df = pd.DataFrame(results)

# –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
print("\n –ì–æ—Ç–æ–≤–æ! –†–µ–∑—É–ª—å—Ç–∞—Ç:")
if not df.empty:
    df_display = df[['title', 'date', 'url']].copy()
    df_display['text_preview'] = df['text'].str[:100] + "..."
    display(df_display)
else:
    print("–ù–æ–≤–æ—Å—Ç–µ–π –Ω–µ –Ω–∞–π–¥–µ–Ω–æ —Å —Ç–µ–∫—É—â–∏–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏.")


# –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –≤ CSV (–º–æ–∂–Ω–æ —Å–∫–∞—á–∞—Ç—å —á–µ—Ä–µ–∑ –∑–Ω–∞—á–æ–∫ —Å–ª–µ–≤–∞ –≤ Colab)
if not df.empty:
    df.to_csv("ria_news.csv", index=False, encoding='utf-8-sig')
    print("\n –§–∞–π–ª —Å–æ—Ö—Ä–∞–Ω—ë–Ω: ria_news.csv")
else:
    print("\n DataFrame –ø—É—Å—Ç, —Ñ–∞–π–ª CSV –Ω–µ —Å–æ–∑–¥–∞–Ω.")
```

```python

!pip install requests beautifulsoup4 pandas lxml

import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime, timedelta
import time
import re

# –ù–∞—Å—Ç—Ä–æ–π–∫–∏
BASE_URL = 'https://ria.ru'
HEADERS = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0 Safari/537.36'
}

def parse_ria_date(date_str):
    """–ü–∞—Ä—Å–∏–Ω–≥ –¥–∞—Ç—ã: –ø–æ–¥–¥–µ—Ä–∂–∫–∞ '–°–µ–≥–æ–¥–Ω—è', '–í—á–µ—Ä–∞', '5 –Ω–æ—è–±—Ä—è 2025' –∏ —Ç.–¥."""
    if not date_str:
        return None
    now = datetime.now()
    date_str = date_str.strip().replace('\xa0', ' ')

    if '–°–µ–≥–æ–¥–Ω—è' in date_str:
        time_part = re.search(r'\d{1,2}:\d{2}', date_str)
        if time_part:
            h, m = map(int, time_part.group().split(':'))
            return now.replace(hour=h, minute=m, second=0, microsecond=0)
        return now
    elif '–í—á–µ—Ä–∞' in date_str:
        time_part = re.search(r'\d{1,2}:\d{2}', date_str)
        if time_part:
            h, m = map(int, time_part.group().split(':'))
            return (now - timedelta(days=1)).replace(hour=h, minute=m, second=0, microsecond=0)
        return now - timedelta(days=1)
    else:
        months = {
            '—è–Ω–≤–∞—Ä—è': '01', '—Ñ–µ–≤—Ä–∞–ª—è': '02', '–º–∞—Ä—Ç–∞': '03', '–∞–ø—Ä–µ–ª—è': '04',
            '–º–∞—è': '05', '–∏—é–Ω—è': '06', '–∏—é–ª—è': '07', '–∞–≤–≥—É—Å—Ç–∞': '08',
            '—Å–µ–Ω—Ç—è–±—Ä—è': '09', '–æ–∫—Ç—è–±—Ä—è': '10', '–Ω–æ—è–±—Ä—è': '11', '–¥–µ–∫–∞–±—Ä—è': '12'
        }
        for ru, num in months.items():
            date_str = date_str.replace(ru, num)
        try:
            # –£–±–∏—Ä–∞–µ–º –≤–æ–∑–º–æ–∂–Ω—ã–µ "–ú–°–ö" –∏ –ø—Ä–æ—á–µ–µ
            date_str = re.sub(r'[^\d\s:]', ' ', date_str)
            date_str = re.sub(r'\s+', ' ', date_str).strip()
            parts = date_str.split()
            if len(parts) >= 3:
                day, month, year = parts[0], parts[1], parts[2]
                hour, minute = ('00', '00')
                if len(parts) >= 5:
                    hour, minute = parts[3], parts[4]
                dt_str = f"{day} {month} {year} {hour}:{minute}"
                return datetime.strptime(dt_str, "%d %m %Y %H:%M")
            else:
                # –ï—Å–ª–∏ –Ω–µ—Ç –≥–æ–¥–∞ ‚Äî –¥–æ–±–∞–≤–ª—è–µ–º —Ç–µ–∫—É—â–∏–π
                day, month = parts[0], parts[1]
                hour, minute = ('00', '00')
                if len(parts) >= 4:
                    hour, minute = parts[2], parts[3]
                dt_str = f"{day} {month} {now.year} {hour}:{minute}"
                return datetime.strptime(dt_str, "%d %m %Y %H:%M")
        except Exception as e:
            print(f"–û—à–∏–±–∫–∞ –ø–∞—Ä—Å–∏–Ω–≥–∞ –¥–∞—Ç—ã '{date_str}': {e}")
            return None

def get_news_list(max_news=None, since_date=None):
    """–ü–∞—Ä—Å–∏—Ç —Å–ø–∏—Å–æ–∫ –Ω–æ–≤–æ—Å—Ç–µ–π —Å –≥–ª–∞–≤–Ω–æ–π —Å—Ç—Ä–∞–Ω–∏—Ü—ã RIA.RU (–∞–∫—Ç—É–∞–ª—å–Ω–æ –Ω–∞ –Ω–æ—è–±—Ä—å 2025)."""
    print("–ó–∞–≥—Ä—É–∑–∫–∞ –≥–ª–∞–≤–Ω–æ–π —Å—Ç—Ä–∞–Ω–∏—Ü—ã RIA.RU...")
    response = requests.get(BASE_URL, headers=HEADERS)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'lxml')

    # –ù–æ–≤—ã–µ —Å–µ–ª–µ–∫—Ç–æ—Ä—ã: –∏—â–µ–º –≤—Å–µ —Å—Å—ã–ª–∫–∏ –≤–Ω—É—Ç—Ä–∏ article –∏–ª–∏ —Å –∫–ª–∞—Å—Å–æ–º, –ø–æ—Ö–æ–∂–∏–º –Ω–∞ –Ω–æ–≤–æ—Å—Ç—å
    news_links = soup.select('a[href^="/"]:has(span), a[href^="https://ria.ru/"]:has(span)')

    if not news_links:
        # –†–µ–∑–µ—Ä–≤–Ω—ã–π –≤–∞—Ä–∏–∞–Ω—Ç: –≤—Å–µ —Å—Å—ã–ª–∫–∏ —Å / –∏ –¥–ª–∏–Ω–æ–π > 20
        news_links = [a for a in soup.find_all('a', href=True)
                      if a['href'].startswith('/') and len(a['href']) > 20 and '/tags/' not in a['href']]

    print(f"–ù–∞–π–¥–µ–Ω–æ –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω—ã—Ö –Ω–æ–≤–æ—Å—Ç–µ–π: {len(news_links)}")

    parsed = []
    for a in news_links:
        if max_news and len(parsed) >= max_news:
            break

        title_tag = a.find('span')
        if not title_tag:
            continue
        title = title_tag.get_text(strip=True)
        if not title or len(title) < 10:  # –ø—Ä–æ–ø—É—Å–∫–∞–µ–º –∫–æ—Ä–æ—Ç–∫–∏–µ –∑–∞–≥–æ–ª–æ–≤–∫–∏
            continue

        href = a['href']
        url = href if href.startswith('http') else BASE_URL + href

        # –ò—â–µ–º –¥–∞—Ç—É ‚Äî —á–∞—Å—Ç–æ —Ä—è–¥–æ–º —Å –∑–∞–≥–æ–ª–æ–≤–∫–æ–º –∏–ª–∏ –≤ —Ä–æ–¥–∏—Ç–µ–ª—å—Å–∫–æ–º –±–ª–æ–∫–µ
        date_str = None
        date_candidate = a.find_next(string=re.compile(r'(–°–µ–≥–æ–¥–Ω—è|–í—á–µ—Ä–∞|\d+\s+\w+)'))
        if date_candidate:
            date_str = date_candidate.strip()
        else:
            # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ä–æ–¥–∏—Ç–µ–ª—å—Å–∫–∏–µ —ç–ª–µ–º–µ–Ω—Ç—ã
            parent = a.find_parent()
            if parent:
                date_elem = parent.find(string=re.compile(r'(–°–µ–≥–æ–¥–Ω—è|–í—á–µ—Ä–∞|\d+\s+\w+)'))
                if date_elem:
                    date_str = date_elem.strip()

        pub_date = parse_ria_date(date_str)

        # –§–∏–ª—å—Ç—Ä –ø–æ –¥–∞—Ç–µ
        if since_date and pub_date and pub_date < since_date:
            if max_news is None:
                break

        parsed.append({
            'title': title,
            'url': url,
            'date_str': date_str,
            'date': pub_date
        })

    print(f"–û—Ç–æ–±—Ä–∞–Ω–æ –Ω–æ–≤–æ—Å—Ç–µ–π: {len(parsed)}")
    return parsed

def get_article_text(url):
    """–ü–∞—Ä—Å–∏–Ω–≥ —Ç–µ–∫—Å—Ç–∞ —Å—Ç–∞—Ç—å–∏ –Ω–∞ RIA.RU."""
    try:
        resp = requests.get(url, headers=HEADERS, timeout=10)
        soup = BeautifulSoup(resp.text, 'lxml')
        # –¢–µ–∫—Å—Ç —Å—Ç–∞—Ç—å–∏ ‚Äî –≤ <div class="article__body"] –∏–ª–∏ <div class="content"]
        text_blocks = soup.select('div.article__body > div:not([class])') or \
                      soup.select('div.article__text') or \
                      soup.select('div[data-type="text"]') or \
                      soup.select('div.content > p')

        if not text_blocks:
            # –ï—Å–ª–∏ –Ω–µ –Ω–∞—à–ª–∏ ‚Äî –ø–æ–ø—Ä–æ–±—É–µ–º –≤—Å–µ –ø–∞—Ä–∞–≥—Ä–∞—Ñ—ã –≤ –æ—Å–Ω–æ–≤–Ω–æ–º –∫–æ–Ω—Ç–µ–Ω—Ç–µ
            main_content = soup.select_one('div.article__body, div.content')
            if main_content:
                text_blocks = main_content.find_all('p')

        text = '\n'.join([p.get_text(strip=True) for p in text_blocks if p.get_text(strip=True)])
        return text if text else "–¢–µ–∫—Å—Ç –Ω–µ –Ω–∞–π–¥–µ–Ω"
    except Exception as e:
        return f"–û—à–∏–±–∫–∞ –∑–∞–≥—Ä—É–∑–∫–∏ —Ç–µ–∫—Å—Ç–∞: {str(e)}"

MAX_NEWS = 13      # –°–∫–æ–ª—å–∫–æ –Ω–æ–≤–æ—Å—Ç–µ–π –ø–∞—Ä—Å–∏—Ç—å? (None ‚Äî –µ—Å–ª–∏ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç–µ –¥–∞—Ç—É)
SINCE_DATE_STR = None # –ü—Ä–∏–º–µ—Ä: "2025-11-01" (None ‚Äî –µ—Å–ª–∏ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç–µ MAX_NEWS)

#  –∑–∞–ø—É—Å–∫

since_date = datetime.strptime(SINCE_DATE_STR, "%Y-%m-%d") if SINCE_DATE_STR else None

if SINCE_DATE_STR:
    print(f"–ü–∞—Ä—Å–∏–Ω–≥ –Ω–æ–≤–æ—Å—Ç–µ–π —Å {SINCE_DATE_STR}")
else:
    print(f"–ü–∞—Ä—Å–∏–Ω–≥ {MAX_NEWS} –ø–æ—Å–ª–µ–¥–Ω–∏—Ö –Ω–æ–≤–æ—Å—Ç–µ–π")

news_list = get_news_list(max_news=MAX_NEWS, since_date=since_date)

results = []
for i, news in enumerate(news_list, 1):
    print(f"[{i}/{len(news_list)}] {news['title'][:60]}...")
    text = get_article_text(news['url'])
    results.append({
        'title': news['title'],
        'url': news['url'],
        'date': news['date'],
        'text': text
    })
    time.sleep(0.8)

# –°–æ–∑–¥–∞–Ω–∏–µ DataFrame
df = pd.DataFrame(results)

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
print("\n –†–µ–∑—É–ª—å—Ç–∞—Ç:")
display(df[['title', 'date', 'url']].fillna(''))

# –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ
df.to_csv("ria_news.csv", index=False, encoding='utf-8-sig')
print("\nüìÅ –§–∞–π–ª ria_news.csv –≥–æ—Ç–æ–≤ –∫ —Å–∫–∞—á–∏–≤–∞–Ω–∏—é (–ª–µ–≤–∞—è –ø–∞–Ω–µ–ª—å ‚Üí Files)")
```
